# 당장 좋은 것만 선택하는 그리디

- 그리디는 단순하지만 강력한 문제 해결 방법
- 탐욕법, 탐욕적으로 문제 푸는 알고리즘
  - 탐욕적: 현재 상황에서 지금 당장 좋은 것만 고르는 방법
  - 매 순간 가장 좋은 것을 선택
  - 현재의 선택이 나중에 미칠 영향을 고려하지 않음.
- 그리디는 사전에 외우고 있지 않아도 풀 수 있을 가능성이 높은 문제
- 그 외 정렬, 최단 경로 등의 알고리즘은 사용 방법을 정확히 알고 있어야만 해결 가능
- 다익스트라 알고리즘은 그리디 알고리즘
- 그리디는 암기가 필요한 알고리즘
- 그리디 알고리즘 자체가 문제 문제 출제의 폭이 넓기 때문에
  - 다익스트라 알고리즘과 같은 특이 케이스를 제외하고는
  - 단순 암기로 모든 문제 대처가 어려움
- 많은 유형을 접해보고 문제를 풀어보며 훈련해야 한다.
- 그리디는 창의력, 즉 문제를 풀기 위한 최소한의 아이디어를 떠올릴 수 있는 능력 요구
- 특정 문제를 만났을 때 단순히 현재 상황에서 가장 좋아 보이는 것만을 선택해도 문제를 풀 수 있는지 파악
- 그리디는 기준에 따라 좋은 것을 선택하므로
  - 문제에서 "가장 큰 순서대로", "가장 작은 순서대로"와 같은 기준을 제시
  - 대체로 이 기준은 정렬 알고리즘을 사용했을 때 만족시킬 수 있으므로
  - 그리디는 정렬 알고리즘과 짝을 이뤄 출제된다.

## 예시 - 거스름돈

- 당신은 계산을 도와주는 점원이다.
- 카운터에는 거스름돈으로 500원, 100원, 50원, 10원짜리 동전이 무한히 존재한다.
- 손님에게 거슬러줘야 할 돈이 N원일 때 거슬러줘야 할 동전의 최소 개수를 구하라.
- 단, 거슬러 줘야 할 돈 N은 항상 10의 배수이다.

```python
def solution(rest):
    count = 0
    coin_types = [500, 100, 50, 10]

    for coin in coin_types:
        count += rest // coin # count == 500원 2개, 100원 2개, 50원 1개, 10원 1개
        rest = rest % coin # rest == 260, 60, 10, 0

    return count

print(solution(1260))
```

### 문제 해설
- 동전의 최소 개수이므로, 가장 큰 화폐 단위부터 돈을 거슬러 준다.
- 코드를 보면 화폐의 종류만큼 반복 수행한다.
    - 따라서 화폐의 종류가 K개라면 시간 복잡도는 O(K)가 된다.
    - 시간 복잡도에서 거슬러 줘야 할 돈 N은 없다.
    - 즉, 이 알고리즘의 시간 복잡도는
        - 동전의 총 종류에만 영향을 받는다.
        - 거슬러 줘야할 금액의 크기와는 무관하다.

### 그리디 알고리즘의 정당성
- 모든 알고리즘에 적용할 수 있는 문제는 아니다.
- 대부분의 문제는 그리디 알고리즘을 이용했을 때 '최적의 해'를 찾을 수 없을 가능성이 다분하다.
- 하지만 거스름돈 문제에서 '가장 큰 화폐 단위부터' 돈을 거슬러 주는 것과 같이,
- **탐욕적으로 문제에 접근했을 때 정확한 답을 찾을 수 있다는 보장**이 있을 때 매우 효과적이고 직관적이다.

- 그리디 알고리즘으로 해법을 찾았을 때는 그 해법이 정당한지 검토해야 한다.
- 거스름돈 문제를 그리디로 해결할 수 있는 이유는
    - 가지고 있는 동전 중에 큰 단위가 항상 작은 단위의 배수이므로, 작은 단위의 동전들을 종합해 다른 해가 나올 수 없기 때문이다.
- 예를 들어
    - 거스름돈 800원, 화폐 단위 500원, 400원, 100원이라면
    - 그리디 알고리즘으로는 4개의 동전(500 + 100 + 100 + 100)
    - 실제로 최적의 해는 2개의 동전(400 + 400)이다.
- 다시 말해 이 문제는 큰 단위가 작은 단위의 배수 형태이므로,
    - 큰 단위의 화폐부터 작은 단위의 화폐까지 차례대로 확인해서 거슬러 주는 작업만 수행하면 된다는 아이디어가 정당하다.

- 대부분의 그리디 문제는 이처럼 문제 풀이를 위한 최소한의 아이디어를 떠올리고, 정당한지 검토할 수 있어야 답 도출 가능

- 어떤 문제를 만났을 때, 유형 파악이 어렵다면, 탐욕적 해결이 가능한지 고민해보자.
- 만약 그리디로 해결 방법을 찾을 수 없다면, 아래 방법으로 해결 가능한지 재차 고민하는 것도 방법이다.
    - 다이나믹 프로그래밍
    - 그래프 알고리즘 등
- 처음 문제를 만났을 때는 다양한 아이디어를 고려해야 한다.
    1. 가장 먼저 10원으로 모두 거슬러 준다면 어떻게 되지?
        - 10원으로 모두 거슬러 주면 최적의 해를 구할 수 없겠다는 문제점 인식
    2. 가장 큰 500원부터 차례대로 거슬러 준다면 어떻게 되지?
        - 큰 단위가 항상 작은 단위의 배수 형태이므로, 최적의 해를 보장할 수 있겠다는 정답 판정

- 알고리즘 대회 및 코딩 테스트에 출제되었던 문제
    - 기업의 코딩 테스트를 안정적으로 통과하려면
    - 각 문제별 아이디어를 떠올리고, 코드로 작성하기까지 30분 이내로 소요되어야 한다.

## 실전 - 큰 수의 법칙

- 동빈이는 자신만의 방식으로 사용하고 있다.
- 다양한 수로 이루어진 배열이 있을 때, 주어진 수를 M번 더하여 가장 큰 수를 만드는 법칙이다.
- 단, 배열의 특정 인덱스에 해당하는 수가 연속해서 K번을 초과하여 더해질 수 없다.

- 예시 1
    - M = 8, K = 3, N = [2, 4, 5, 4, 6]
    - 특정 인덱스의 수가 연속해서 세 번까지만 더해질 수 있으므로
    - 6 + 6 + 6 + 5 + 6 + 6 + 6 + 5 = 46이 정답

- 예시 2
    - M = 7, K = 2, N = [3, 4, 3, 4, 3]
    - 가장 큰 원소가 2개이므로, 번갈아 K번까지 더하는 것이 가능하다.
    - 4 + 4 + 4 + 4 + 4 + 4 + 4 = 28이 정답

- 입력 조건
    - 2 <= N < 1000
    - 1 <= M < 10000
    - 1 <= K < 10000
    - K는 항상 M보다 작거나 같다.

```python
first_line = "5 8 3" # N M K (M은 반드시 K보다 크다.)
second_line = "2 4 5 4 6" # N개의 자연수, 1 이상 10,000 이하

def solution(first_line, second_line):
    # 값 가져오기
    n, m, k = map(int, first_line.split())
    data = list(map(int, second_line.split()))

    data.sort(reverse=True)
    first = data[0]
    second = data[1]

    result = 0

    # 방법 1
    # count = m
    # while True:
    #     # first를 k 전까지 계산하기
    #     for _ in range(k):
    #         if count == 0: # 이 과정을 count까지 수행
    #             break
    #         result += first
    #         count -= 1
        
    #     if count == 0: # 이 과정을 count까지 수행
    #         break
        
    #     # first를 k를 초과했을 경우, second로 계산하기
    #     result += second
    #     count -= 1
    
    # 방법 2
    count = int(m / (k+1)) * k
    count += m % (k+1)

    result += count * first
    result += (m - count) * second

    return result

print(solution(first_line, second_line))
```

### 문제 해설
- 이 문제는 전형적인 그리디 문제로, 아이디어를 떠올리는 것은 어렵지 않다.
- 다만, 구현 실수로 인해 오답 처리를 받는 경우가 많은 문제이다.
- 꼭 직접 코드를 작성해보는 것을 권장한다.
- 가장 큰 수와 두 번째로 큰 수만 저장하면 된다.
- 가장 큰 수(A)를 K번 더하고 두 번째로 큰 수(B)를 한 번 더하면 된다.

- 방법 1
    - 반복문을 이용해서
    - A를 K번 더하고 B를 한 번 더하면 된다.
    - M이 10000 이하이므로, 이 방식으로 문제 해결 가능하지만,
    - (한계) M이 100억 이상으로 커진다면, 시간 초과 판정이 될 것

- (해결) 방법 2
    - 간단한 수학적 아이디어로 더 효율적으로 문제 해결 가능
    - A와 B가 더해질 때는 일정한 패턴으로 반복되어 더해진다.
    - 반복되는 횟수(수열의 길이)는 M을 (K+1)로 나눈 몫이 된다.
    - 다만, 이 때 M이 (K+1)로 나누어떨어지지 않는 경우가 있다.
    - 나머지만큼 A가 더해지므로 이를 고려한다.
    - $count = int(M / (K + 1)) * K + M \% (K + 1)$
    - 따라서 count만큼 A를 곱셈, M-count만큼 B를 곱셈하여 더하면 가장 큰 수가 된다.

## 실전 - 숫자 카드 게임

- 숫자 카드 게임은 여러 개의 숫자 카드 중에서 가장 높은 숫자가 쓰인 카드 한 장을 뽑는 게임이다.
- 게임의 룰은 아래와 같다.
    1. 숫자가 쓰인 카드들이 N * M 형태로 놓여 있다. N은 행, M은 열의 개수를 의미한다.
    2. 먼저 뽑고자 하는 카드가 포함되어 있는 행을 선택한다.
    3. 그다음 선택된 행이 포함된 카드들 중 가장 숫자가 낮은 카드를 뽑아야 한다.
    4. 따라서 처음에 행을 선택할 때, 이후 해당 행에서 가장 숫자가 낮은 카드를 뽑을 것을 고려하여 최종적으로 가장 높은 숫자를 뽑을 수 있도록 전략을 세운다.

- 입력 조건
    - 첫째 줄, N >= 1, M <= 100
    - 둘째 줄, N개의 줄에 걸쳐 각 카드에 적힌 숫자가 주어진다.
    - 각 카드에 적힌 숫자 <= 10000

- 입력으로 들어오는 수는 모두 10,000 이하이므로 단순히 배열에서 가장 작은 숫를 찾고, 그 수 중에서 가장 큰 수를 찾는 방식으로 해결한다.

## 3. 실전 - 1이 될 때까지

- 어떤 수 N이 1이 될 때까지 다음 두 과정 중 하나를 반복적으로 선택하여 수행하려고 한다.
- 단, 두 번째 연산은 N이 K로 나누어떨어질 때만 선택할 수 있다.
    1. N에서 1을 뺀다.
    2. N을 K로 나눈다.
- N, K가 주어질 때 N이 1이 될때까지 반복해야 될 최소 횟수를 구하라.

```python
"""
    25 5
"""
n, k = map(int, input().split())

count = 0

# 방법 1
# while n != 1:
#     if n % k == 0:
#         n = n / k
#     else:
#         n -= 1
#     count += 1

# 방법 2
# while True:
#     # (N == K로 나누어 떨어지는 수)가 될 때까지 1씩 뺴기
#     target = (n // k) * k # N에 가장 가까운 K의 배수
#     count += (n - target) # 빼야 할 횟수를 한 번에 더하기
#     n = target

#     # N이 K보다 작을 때(더 이상 나눌 수 없을 때) 반복문 탈출
#     if n < k:
#         break

#     # K로 나누기
#     count += 1
#     n //= k

# 방법 3 (AI 추천)
while n >= k:
    count += n % k + 1
    n //= k

# 마지막으로 남은 수에 대하여 1씩 뺴기
count += (n - 1)
print(count)
```

### 문제 해설
