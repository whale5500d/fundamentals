# 최단 경로

## 가장 빠르게 도달하는 방법

- 최단 경로(Shortest Path) 알고리즘은 가장 짧은 경로를 찾는 알고리즘
- 길 찾기 문제라고도 불림
- 최단 경로 알고리즘에는 다양한 유형이 있는데, 상황에 맞게 효율적인 알고리즘이 정립되어 있음

- 예를 들어 '한 지점에서 다른 특정 지점까지 최단 경로를 구해야 하는 경우', '모든 지점에서 다른 모든 지점까지 최단 경로를 모두 구해야 하는 경우' 등
- 사례에 맞는 알고리즘을 알고 있다면 문제를 쉽게 풀 수 있음.

- 최단 경로 문제는 그래프를 이용해 표현하는데 각 지점은 그래프에서 '노드'로 표현되고, 지점 간 연결된 도로는 그래프에서 '간선'으로 표현.
- 또한 실제 코딩 테스트에서는 최단 경로를 모두 출력하는 문제보다 단순한 최단 거리를 출력하도록 요구하는 문제가 많이 출제됨

- 컴퓨터공학과 학부 수준에서 사용하는 최단 거리 알고리즘은 아래 3가지
    - 다익스트라 최단 경로 알고리즘
    - 플로이드 워셜
    - 벨만 포드 알고리즘
- 그 중 다익스트라 최단 경로와 플로이드 워셜 알고리즘 유형을 다룰 예정. 코딩 테스트에서 가장 많이 등장하기 때문
- 앞서 학습한 그리디 알고리즘, 다이나믹 프로그래밍 알고리즘이 최단 경로 알고리즘에 그대로 적용됨
- 여기서 배우는 내용은 그리디 알고리즘, 다이나믹 프로그래밍 알고리즘의 한 유형

## 다익스트라 최단 경로 알고리즘

- 다익스트라(Dijkstra) 최단 경로 알고리즘은 그래프에서 여러 개의 노드가 있을 때, 특정 노드에서 출발하여 다른 노드로 가는 각각의 최단 경로를 구해주는 알고리즘

- 다익스트라 최단 경로 알고리즘은 기본적으로 그리디 알고리즘으로 분류. 매번 '가장 비용이 적은 노드'를 선택해서 임의의 과정을 반복하기 때문.
- 다익스트라 최단 경로 알고리즘의 원리
    1. 출발 노드 설정
    2. 최단 거리 테이블을 초기화
    3. 방문하지 않은 노드 중 최단 거리가 가장 짧은 노드를 선택
    4. 해당 노드를 거쳐 다른 노드로 가는 비용을 계산하여 최단 거리 테이블 갱신
    5. 3, 4번 반복

- 다익스트라 최단 경로 알고리즘 구현 방법 2가지
    - 구현하기 쉽지만 느리게 동작하는 코드
    - 구현하기 까다롭지만 빠르게 동작하는 코드

- 선택된 노드는 '최단 거리'가 완전히 선택된 노드이므로, 더 이상 알고리즘을 반복해도 최단 거리가 줄어들지 않음. 다시 말해 다익스트라 알고리즘이 진행되면서 **한 단계당 하나의 노드에 대한 최단 거리를 확실히 찾는 방법**

### 방법 1. 간단한 다익스트라 알고리즘

- O(V^2)의 시간 복잡도를 가짐. V는 노드의 개수를 의미함.
- 이 알고리즘은 직관적이고 쉽게 이해할 수 있음.
- 처음에 각 노드에 대한 최단 거리를 담는 1차원 리스트를 선언
- 이후에 단계마다 '방문하지 않은 노드 중 최단 거리가 가장 짧은 노드를 선택'하기 위해 매 단계마다 1차원 리스트의 모든 원소를 확인(순차 탐색)

```python
# 간단한 다익스트라 알고리즘 소스코드
# import sys
# input = sys.stdin.readline
input_data = """
6 11
1
1 2 2
1 3 5
1 4 1
2 3 3
2 4 2
3 2 3
3 6 5
4 3 3
4 5 1
5 3 1
5 6 2
""".strip().splitlines()
input_index = 0
def input():
    global input_index
    result = input_data[input_index]
    input_index += 1
    return result


INF = int(1e9) # 1_000_000_000
n, m = map(int, input().split())


# 출발 노드 설정
start = int(input())

# 최단 거리 테이블 초기화
graph = [[] for i in range(n+1)]
visited = [False] * (n+1)
distance = [INF] * (n+1)

for _ in range(m):
    u, v, w = map(int, input().split()) # u는 출발 노드, v는 도착 노드, w는 weight(가중치)
    graph[u].append((v, w))

# 방문하지 않은 노드를 거쳐 최단 거리가 가장 짧은 노드 (인덱스) 선택
def get_smallest_node():
    min_value = INF
    index = 0
    for i in range(1, n+1):
        if distance[i] < min_value and not visited[i]:
            min_value = distance[i]
            index = i
    return index

# 해당 노드를 거쳐 다른 노드로 가는 가장 짧은 비용을 계산하여 최단 거리 테이블 갱신
def dijkstra(start):
    distance[start] = 0
    visited[start] = True

    for j in graph[start]:
        distance[j[0]] = j[1]

    # 반복
    for _ in range(n-1):
        now = get_smallest_node()
        visited[now] = True

        for j in graph[now]:
            cost = distance[now] + j[1]
            if cost < distance[j[0]]:
                distance[j[0]] = cost

dijkstra(start)

# 출력
for i in range(1, n+1):
    if distance[i] == INF:
        print("INFINITY")
    else:
        print(f"distance {i}: {distance[i]}")
```

### 문제 해설
- 입력되는 데이터가 많다는 가정 하에 input() 대신 sys.std.readline()으로 치환하여 사용 예정
- DFS, BFS에서 소스코드와 마찬가지로 모든 리스트는 (노드의 개수+1)의 크기로 할당하여, 노드의 번호를 인덱스로 하여 바로 리스트에 접근할 수 있도록 설계함. 그래프를 표현해야 할 때 많이 사용하는 일반적인 코드 작성법

### 간단한 다익스트라 알고리즘의 시간 복잡도
- 시간 복잡도는 O(V^2)
- 총 O(V)번에 걸쳐서 최단 거리가 가장 짧은 노드를 매번 선형 탐색해야 하고, 현재 노드와 연결된 노드를 매번 일일이 확인하기 때문
- 따라서 코딩 테스트의 최단 경로 문제에서 전체 노드의 개수가 5,000개 이하면 일반적으로 해당 코드로 푸는데 무리가 없음.

- 하지만 노드의 개수가 10,000개를 넘어가는 문제라면 해당 코드로 해결하기 어려움. 노드의 개수 및 간선의 개수가 많을 때는 '개선된 다익스트라 알고리즘'을 이용해야 함.

### 방법 2. 개선된 다익스트라 알고리즘

- 다익스트라 알고리즘을 간단히 구현하면 시간 복잡도가 O(V^2)
- 개선된 다익스트라 알고리즘을 사용하면 최악의 경우에도 시간 복잡도 O(ElogV)를 보장
- V는 노드 개수이고, E는 간선 개수이다.
- 최단 거리가 가장 짧은 노드를 찾기 위해, 매번 최단 거리 테이블을 선형적으로(모든 원소를 앞에서부터 하나씩) 탐색해야 했다.
- 개선된 다익스트라 알고리즘에서는 힙(Heap) 자료구조를 사용한다.

Step 0
1. 출발 노드를 제외한 모든 노드의 최단 거리를 무한으로 설정
2. 이후 우선순위 큐에 1번 노드를 넣는다.
3. 1번 노드로 가는 거리는 자기 자신까지 도달하는 거리이기 때문에 0이다. (거리: 0, 노드: 1)의 정보의 객체를 우선순위 큐에 넣는다. 원소로 튜플을 입력받으면, 튜플의 첫 번째 원소를 기준으로 우선순위 큐를 구성한다.

Step 1
1. 거리가 짧은 원소가 우선순위 큐의 최상위 원소로 위치한다.
2. 따라서 우선순위 큐에서 노드를 꺼낸 뒤 해당 노드를 이미 처리한 적이 있다면 무시하고, 아직 처리하지 않은 노드는 처리하면 된다.
3. 현재 노드 번호 1번의 거리만 0이고, 나머지는 무한이므로 우선순위 큐에서는 (0, 1)이 나온다.
4. 이는 1번 노드까지 가는 최단거리가 0이라는 의미이므로, 1번 노드를 거쳐서 2번, 3번, 4번 노드로 가는 최소 비용을 계산한다. (차례대로 2(0+2), 5(0+5), 1(0+1)이다.)
5. 2번, 3번, 4번 노드로 가는 더 짧은 경로를 찾았으므로, '무한'에서 각각 갱신한다.
6. 만들어진 노드 정보는 다시 우선순위 큐에 넣는다.

Step 2
1. 다시 우선순위 큐에서 원소를 꺼내 동일한 과정을 반복한다.
2. 이번에는 (1, 4)의 값을 갖는 원소가 추출된다.
3. 아직 노드 4를 방문하지 않았으며, 현재 최단 거리가 가장 짧은 노드다.
4. 따라서 노드 4를 기준으로 노드 4와 연결된 간선들을 확인한다.
5. 이때 4번 노드까지의 최단 거리는 1이므로 4번 노드를 거쳐서 3번과 5번 노드로 가는 최소 비용을 계산한다. (차례대로 4(1+3), 2(1+1))
6. 이는 기존 리스트에 담겨 있던 값들보다 작기 때문에 리스트가 새로 갱신되고, 우선순위 큐에는 (4, 3), (2, 5)라는 두 원소가 추가로 들어가게 된다.
7. 현재 그림에서는 튜플의 첫 번째 원소(거리)가 작은 순서대로 왼쪽부터 기록하고 있다. 따라서 갱신된 우선순위 큐 또한 그림처럼 그려진다.

Step 3
1. 마찬가지로 Step 3에서는 노드 2에 대해 처리한다.
2. 2번과 5번 노드까지의 최단 거리가 모두 값이 2로 같으므로 어떤 원소부터 처리해도 상관없지만, 우선순위 큐에서 2번 노드가 꺼내졌다고 가정하자.
3. 마찬가지로 2번 노드를 거쳐서 도달할 수 있는 노드 중에서 더 거리가 짧은 경우가 있는지 확인한다.
4. 이번 단계에서는 2번 노드를 거쳐서 가는 경우 중 현재의 최단 거리를 더 짧게 갱신할 수 있는 방법은 없다.
5. 따라서 우선순위 큐에 어떠한 원소도 들어가지 않고 리스트가 갱신된다.

Step 4
1. Step 4에서는 노드 5에 대해 처리한다.
2. 5번 노드를 거쳐서 3번, 6번 노드로 갈 수 있다. 현재 5번 노드까지 가는 최단 거리가 2이므로 5번 노드에서 3번 노드로 가는 거리인 1을 더한 3이 기존의 값인 4보다 작다.
3. 따라서 3번 노드는 4에서 새로운 값인 3으로 갱신한다.
4. 또한 6번 노드로 가는 최단 거리 역시 마찬가지로 갱신된다.

Step 5
1. 마찬가지로 원소 (3, 3)을 꺼내서 3번 노드를 기준으로 알고리즘을 수행한다.
2. 최댄 거리 테이블이 갱신되지 않는다.

Step 6
1. 이어서 원소 (4, 3)을 꺼내서 3번 노드를 기준으로 알고리즘을 수행한다.
2. 다만, 3번 노드는 앞서 처리된 적이 있다.
3. 현재 우선순위 큐에서 꺼낸 원소에는 3번 노드까지 최단 거리가 4라는 정보가 들어 있다. 따라서 현재 노드인 3번에 대해서 이미 처리된 것으로 볼 수 있으므로 현재 우선순위 큐에서 꺼낸 (4, 3)이라는 원소는 무시하면 된다.

Step 7
1. 이어서 원소 (4, 6)이 꺼내진다. 따라서 6번 노드에 대해서 처리한다.

Step 8
1. 마지막으로 남은 원소를 꺼낸다.
2. 이미 처리된 노드이므로 무시한다.

- 모든 단계를 거친 후에는 최단 거리 테이블에 남아 있는 0, 2, 3, 1, 2, 4가 각 노드의 최단 거리이다. 위의 방법에서는 최단 거리가 가장 짧은 노드를 찾기 위해 우선순위 큐를 이용하고 있으며, 앞서 보여준 방법과 비교했을 때 훨씬 빠르게 동작한다.
- 앞 코드와 비교했을 때 get_samllest_node()라는 함수를 작성할 필요가 없다. '최단 거리가 가장 짧은 노드'를 선택하는 과정을 다익스트라 최단 경로 함수 안에서 우선순위 큐를 이용하는 방식으로 대체할 수 있기 때문이다.

### 힙 설명

- 힙 자료구조는 우선순위 큐(Priority Queue)를 구현하기 위한 자료구조 중 하나
- 우선순위 큐는 우선순위가 가장 높은 데이터를 가장 먼저 삭제
- PrioriyQueue, heapq를 사용할 수 있는데, 일반적으로 heapq가 더 빠르게 동작
- 우선순위 값을 표현할 때 정수형 자료형의 변수가 사용.
- 우선순위 큐 라이브러리에 데이터의 묶음을 넣으면, 첫 번째 원소를 기준으로 우선순위를 설정
- 우선순위 큐는 내부적으로 최소 힙, 최대 힙을 사용.
    - 최소 힙은 가장 낮은 데이터가 먼저 삭제
    - 최대 힙은 가장 높은 데이터가 먼저 삭제
- 다익스트라 최단 경로 알고리즘에서 비용이 적은 노드를 우선하여 방문하므로 최소 힙 구조를 기반으로 사용 (파이썬 라이브러리는 기본적으로 최소 힙 구조)
- 최소 힙을 최대 힙처럼 사용하기 위해 음수 부호를 붙여서 넣었다가, 꺼낸 다음 다시 음수 부호를 붙여서 원래 값으로 돌리는 방식 사용 가능

```python
# 개선된 다익스트라 알고리즘 소스코드
# import sys
# input = sys.stdin.readline
input_data = """
6 11
1
1 2 2
1 3 5
1 4 1
2 3 3
2 4 2
3 2 3
3 6 5
4 3 3
4 5 1
5 3 1
5 6 2
""".strip().splitlines()
input_index = 0
def input():
    global input_index
    result = input_data[input_index]
    input_index += 1
    return result

import heapq
INF = int(1e9) # 1_000_000_000
n, m = map(int, input().split())


# 출발 노드 설정
start = int(input())

# 최단 거리 테이블 초기화
graph = [[] for i in range(n+1)]
distance = [INF] * (n+1)

for _ in range(m):
    u, v, w = map(int, input().split())
    graph[u].append((v, w))

# 해당 노드를 거쳐 다른 노드로 가는 가장 짧은 비용을 계산하여 최단 거리 테이블 갱신
def dijkstra(start):
    q = []
    heapq.heappush(q, (0, start))
    distance[start] = 0
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist: # lazy deletion (지연 삭제)
            continue
        for i in graph[now]:
            cost = dist + i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))

dijkstra(start)

# 출력
for i in range(1, n+1):
    if distance[i] == INF:
        print("INFINITY")
    else:
        print(f"distance {i}: {distance[i]}")
```

### 코드 해석
- 한 번 처리된 노드는 더 이상 처리되지 않는다. 큐에서 노드를 하나씩 검사하는 반복문은 노드의 개수 V 이상의 횟수로 반복되지 않는다. V번 반복될 때마다 자신과 연결된 간선을 모두 확인한다. 따라서 '현재 우선순위 큐에서 꺼낸 노드와 연결된 다른 노드를 확인'하는 횟수는 최대 간선의 개수(E)만큼 연산이 수행된다.
- 전체 다익스트라 최단 경로 알고리즘은 E개의 원소를 우선순위 큐에 넣었다가 모두 빼내는 연산과 매우 유사하다고 볼 수 있다.

- 다익스트라 최단 경로 알고리즘은 우선순위 큐를 이용한다는 점에서 우선순위 큐를 필요로 하는 다른 문제 유형과도 흡사하다. 그래서 최단 경로를 찾는 문제 외에도 두루 적용된다. 예를 들어 그래프 문제로 유명한 최소 신장 트리 문제를 풀 때에도 일부 알고리즘(Prim 알고리즘)의 구현이 다익스트라 구현과 흡사하다. 다익스트라 알고리즘을 바르게 이해하면, 다른 고급 알고리즘도 이해할 수 있다.

```python
# 우선순위 큐 구현해보기
# binary min-heap(이진 최소 힙)을 배열 기반으로 직접 구현하고
# heappush, heappop을 sift-up, sift-down 연산으로 작성하기

# 단계 1 - 인덱스 헬퍼
def get_parent_index(index):
    return (index - 1) // 2

def get_left_child_index(index):
    return 2 * index + 1

def get_right_child_index(index):
    return 2 * index + 2

# 단계 2 - sift-up, heap_push
def sift_up(heap_list, start_index):
    current_index = start_index
    while current_index > 0: # 루트에 닿으면 종료
        parent_index = get_parent_index(current_index)
        if heap_list[current_index] < heap_list[parent_index]:
            heap_list[current_index], heap_list[parent_index] = \
                heap_list[parent_index], heap_list[current_index]
            current_index = parent_index
        else:
            break # 부모가 더 작으면 불변식 만족

def heap_push(heap_list, item):
    heap_list.append(item) # 끝에 추가 (완전 트리 유지)
    sift_up(heap_list, len(heap_list)-1) # 위로 올려 불변식 복원

# 단계 3 - sift-down과 heap_pop
def sift_down(heap_list, start_index):
    current_index = start_index
    heap_size = len(heap_list)
    while True:
        left_child_index = get_left_child_index(current_index)
        right_child_index = get_right_child_index(current_index)
        smallest_index = current_index

        # 두 자식 중 더 작은 쪽을 찾는다
        if left_child_index < heap_size and heap_list[left_child_index] < heap_list[smallest_index]:
            smallest_index = left_child_index
        if right_child_index < heap_size and heap_list[right_child_index] < heap_list[smallest_index]:
            smallest_index = right_child_index

        if smallest_index == current_index: # 자기가 가장 작으면 종료
            break

        heap_list[current_index], heap_list[smallest_index] = \
            heap_list[smallest_index], heap_list[current_index]
        current_index = smallest_index

def heap_pop(heap_list):
    if not heap_list:
        raise IndexError("pop from empty heap")
    
    last_item = heap_list.pop()
    if not heap_list:
        return last_item
    
    minimum_item = heap_list[0]
    heap_list[0] = last_item
    sift_down(heap_list, 0)
    return minimum_item

# 단계 4 - 검증
import heapq
import random

def is_valid_min_heap(heap_list):
    """모든 부모 <= 자식 인지 전수 검사 (불변식 검증)"""
    for index in range(len(heap_list)):
        for child_index in (get_left_child_index(index), get_right_child_index(index)):
            if child_index < len(heap_list):
                if heap_list[index] > heap_list[child_index]:
                    return False
    return True

def test_against_standard_library():
    my_heap = []
    standard_heap = []

    for _ in range(10):
        if my_heap and random.random() < 0.4:
            print(my_heap)
            assert heap_pop(my_heap) == heapq.heappop(standard_heap)
        else:
            value = random.randint(0, 1000)
            heap_push(my_heap, value)
            heapq.heappush(standard_heap, value)

    assert is_valid_min_heap(my_heap)
    print("표준 라이브러리 교차 검증 통과")

test_against_standard_library()
```

## 플로이드 워셜 알고리즘

- 다익스트라 알고리즘은 '한 지점에서 다른 특정 지점까지 최단 경로를 구해야 하는 경우'에 사용 가능
- 플로이드 워셜 알고리즘은 '모든 지점에서 다른 모든 지점까지의 최단 경로를 모두 구하는 경우'에 사용 가능

- 다익스트라 알고리즘은 단계마다 최단 거리를 가지는 노드를 하나씩 반복적으로 선택한다. 그리고 해당 노드를 거쳐 가는 경로를 확인하며, 최단 거리 테이블을 갱신하는 방식으로 동작한다.
- 플로이드 워셜 알고리즘 또한 단계마다 '거쳐 가는 노드'를 기준으로 알고리즘을 수행한다. 하지만 방문하지 않은 노드 중에 최단 거리를 갖는 노드를 매번 찾을 필요가 없다는 점이 다르다.
- 노드의 개수가 N개일 때 알고리즘 상으로 N번의 단계를 수행하며, 단계마다 O(N^2)의 연산을 통해 '현재 노드를 거쳐 가는' 모든 경로를 고려한다. 따라서 플로이드 워셜 알고리즘의 총시간 복잡도는 O(N^3)이다.

- 다익스트라 알고리즘에서는 출발 노드가 1개이므로 다른 모든 노드까지의 최단 거리를 저장하기 위해서 1차원 리스트를 이용했다. 반면에 플로이드 워셜 알고리즘은 다익스트라 알고리즘과는 다르게 2차원 리스트에 '최단 거리' 정보를 저장한다는 특징이 있다. 모든 노드에 대해 다른 모든 노드로 가는 최단 거리 정보를 담아야 하기 때문이다. 다시 말해 2차원 리스트를 처리해야 하므로 N번의 단계에서 매번 O(N^2)의 시간이 소요된다.
- 또한 다익스트라 알고리즘은 그리드 알고리즘인데, 플로이드 워셜 알고리즘은 다이나믹 프로그래밍이라는 특징이 있다. 노드의 개수가 N이라고 할 때, N번 만큼의 단계를 반복하며 '점화식'에 맞게 2차원 리스트를 갱신하기 때문이다.

- 각 단계에서는 해당 노드를 거쳐 가능 경우를 고려한다. 예를 들어 1번 노드에 대해서 확인할 때는 1번 노드를 중간에 거쳐 지나가는 모든 경우를 고려하면 된다.
- 따라서 알고리즘에서는 현재 확인하고 있는 노드를 제외하고, N-1개의 노드 중에서 서로 다른 노드(A, B)쌍을 선택한다. 이후에 A -> 1번 노드 -> B로 가는 비용을 확인한 뒤 최단 거리를 갱신한다. 다시 말해 $_{N-1}P_2$개의 쌍을 단계마다 반복해서 확인하면 된다. (이 때 O($_{N-1}P_2$)는 O(N^2)이라고 볼 수 있기 떄문에, 전체 시간 복잡도는 O(N^3)이다.)
- 구체적인 (K번의 단계에 대한) 점화식은 아래와 같다.

$$
D_ab = min(D_ab, D_ak + D_kb)
$$
- 따라서 전체적으로 3중 반복문을 이용하여 이 점화식에 따라 최단 거리 테이블을 갱신하면 된다. 위의 점화식이 의미하는 내용을 말로 풀어 설명하면, 'A에서 B로 가는 최소 비용'과 'A에서 K를 거쳐 B로 가는 비용'을 비교하여더 작은 값으로 갱신하겠다는 뜻이다.

### 최종 결과

- 노드의 개수가 4개이므로 총 [step 4]까지 알고리즘을 수행했다. 그래서 [step 4]가 모두 수행되었을 때 테이블의 형태가 있다. 여기 기록되어 있는 내용이 모든 노드에서 모든 노드로 가는 최단 거리 정보를 표현하고 있다.

```python
# 플로이드 워셜 알고리즘 소스코드

input_data = """
    4
    7
    1 2 4
    1 4 6
    2 1 3
    2 3 7
    3 1 5
    3 4 4
    4 3 2
""".strip().splitlines()
input_index = 0
def input():
    global input_index
    result = input_data[input_index]
    input_index += 1
    return result

n = int(input())
m = int(input())

# 초기화 1
INF = int(1e9)
graph = [[INF] * (n+1) for _ in range(n+1)]

# 초기화 2
for a in range(1, n+1):
    for b in range(1, n+1):
        if a == b:
            graph[a][b] = 0

# 초기화 3
for _ in range(m):
    u, v, w = map(int, input().split())
    graph[u][v] = w

# 점화식에 따라 플로이드 워셜 알고리즘 수행
for k in range(1, n+1):
    for a in range(1, n+1):
        for b in range(1, n+1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])

for a in range(1, n+1):
    for b in range(1, n+1):
        if graph[a][b] == INF:
            print("INFINITY", end=" ")
        else:
            print(graph[a][b], end=" ")
    print()
```
