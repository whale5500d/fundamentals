# 탐색 알고리즘 DFS/BFS

## 그래프

- 노드(Node)와 간선(Edge)로 표현
- 이때 노드를 정점(Vertex)라고 말함
- 그래프 탐색이란 하나의 노드를 시작으로 다수의 노드를 방문하는 것을 의미
- 두 노드가 간선으로 연결되어 있다면, "두 노드가 인접하다(Adjacent)"고 표현
- 프로그래밍에서 그래프는 표현 방식 2가지
    - 인접 행렬(Adjacency Matrix): **차원 배열**로 그래프의 연결 관계 표현 방식
    - 인접 리스트(Adjacency List): **리스트**로 그래프의 연결 관계 표현 방식

1. 인접 행렬
    - 2차원 배열에 각 노드가 연결된 형태를 기록하는 방식
        - 다른 언어의 배열을 파이썬에서는 리스트 자료형으로 표현할 수 있으므로 파이썬은 인접 행렬을 리스트로 구현
    - 연결되어 있지 않은 노드는 무한의 비용으로 작성
    - 예시
        ```python
        INF = 999999999

        graph = [
            [0, 7, 5],
            [7, 0, INF],
            [5, INF, 0],
        ]
        ```
2. 인접 리스트
    - 모든 노드에 연결된 노드에 대한 정보를 차례대로 연결하여 저장하는 방식
    - "연결 리스트"라는 자료구조를 이용해 구현
    - 파이썬은 전통적인 프로그래밍 언어에서 배열과 연결 리스트의 기능을 모두 기본으로 제공
        - C++, Java와 같은 프로그래밍 언어에서는 별도로 연결 리스트 기능을 위한 표준 라이브러리 제공
        - 반면 파이썬은 기본 리스트 자료형이 append와 메서드를 제공
    - 파이썬으로 인접 리스트를 이용해 그래프를 표현하고자 할 때 2차원 리스트를 이용하면 가능
    - 예시
        ```python
        graph = [[] for _ in range(3)]

        graph[0].append((1, 7)) # 0 노드에 연결된 노드 1와 거리 7을 저장
        graph[0].append((2, 5))

        graph[1].append((0, 7))

        graph[2].append((0, 5))
        ```

- 두 방식의 차이
    - 코딩 테스트를 위해 학습하는 터라 메모리와 속도 측면에서 살펴보자.
    - 메모리 측면
        - 인접 행렬 방식은 모든 관계를 저장하므로 노드 개수가 많을수록 메모리가 낭비된다.
        - 인접 리스트 방식은 연결된 정보만 저장하므로 메모리를 효율적으로 사용한다.
        - 특정 노드와 연결된 모든 인접 노드를 순회해야 하는경우, 인접 리스트 방식이 인접 행렬 방식에 비해 메모리 공간 낭비가 적다.
    - 속도 측면
        - 이와 같은 속성으로 인접 리스트 방식은 인접 행렬 방식에 비해 특정한 두 노드가 연결되어 있는지 정보를 얻는 속도가 느리다.
        - 인접 리스트 방식은 연결된 데이터를 하났기 확인해야 하기 때문이다.

## DFS

- Depth-First Search
- 깊이 우선 탐색
- 그래프에서 깊은 부분을 우선적으로 탐색하는 알고리즘

- 특정 경로로 탐색하다가 특정 상황에서 최대한 깊숙이 들어가서 노드를 방문한 후, 다시 돌아가 다른 경로로 탐색하는 알고리즘
- 스택 자료구조 사용
- 구체적인 동작 과정
    1. 탐색 시작 노드를 스택에 삽입하고 방문 처리한다.
    2. 스택의 최상단 노드에 방문하지 않은 인접 노드가 있으면 그 인접 노드를 스택에 넣고 방문 처리
    3. 방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드를 꺼낸다.
    4. 2~3번 과정을 더 이상 수행할 수 없을 때까지 반복한다.
- 방문 처리는 스택에 한 번 삽입되어 처리된 노드가 다시 삽입되지 않게 체크하는 것을 의미
    - 방문 처리를 통해 각 노드를 한 번씩만 처리 가능
- DFS의 기능을 생각하면 순서와 상관없이 처리해도 되지만, 코딩 테스트에서는 번호가 낮은 순서부터 처리하도록 명시하는 편이다.
    - 따라서 관행적으로 번호가 낮은 순서부터 처리하도록 구현한다.

````python
# 각 노드가 연결된 정보를 리스트 자료형으로 표현(2차원 리스트)
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7],
]
````

````python
# 각 노드가 방문된 정보를 리스트 자료형으로 표현(1차원 리스트)
visited = [False] * 9

# DFS 메서드 정의
def dfs(graph, vertex, visited):
    visited[vertex] = True # 현재 노드 방문 처리
    print(vertex, end=' ')

    for i in graph[vertex]:
        if not visited[i]:
            dfs(graph, i, visited)

# 함수 호출
dfs(graph, 1, visited)
````

## BFS

- Breadth-First Search
- 너비 우선 탐색
- 그래프에서 가까운 노드부터 탐색하는 알고리즘

- 큐 자료구조 사용
- 구체적인 동작 과정
    1. 탐색 시작 노드를 큐에 삽입하고 방문 처리한다.
    2. 큐에서 노드를 꺼내 해당 노드의 인접 노드 중 방문하지 않은 노드를 모두 큐에 삽입하고 방문 처리한다.
    3. 2번 과정을 더 이상 수행할 수 없을 때까지 반복한다.
- deque 라이브러리를 사용하는 것이 좋으며, 탐색을 수행함에 있어 O(N)의 시간이 소요
- 일반적인 경우 DFS보다 수행 시간이 좋은 편

````python
from collections import deque

# 각 노드가 방문된 정보를 리스트 자료형으로 표현(1차원 리스트)
visited = [False] * 9

# BFS 메서드 정의
def bfs(graph, start, visited):
    queue = [start]
    # queue = deque([start]) # 큐(Queue) 구현을 위해 deque 라이브러리 사용
    visited[start] = True # 현재 노드 방문 처리
    
    while queue:
        vertex = queue.pop(0)
        print(vertex, end=' ')

        for i in graph[vertex]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

bfs(graph, 1, visited)
````

- 코딩 테스트 중 2차원 배열의 탐색 문제를 만나면 그래프 형태로 바꿔서 생각하면 풀이 방법을 쉽게 떠올릴 수 있다.
- 코딩 테스트에서 탐색 문제를 보면 그래프 형태로 표현한 다음 풀이 방법을 고민하자.

## 260605 - 실전 - 음료수 얼려 먹기
- N x M 크기의 얼음 틀이 있다.
- 구멍이 뚫려 있는 부분은 0, 칸막이가 존재하는 부분은 1로 표시
- 구멍이 뚫려 있는 부분끼리 상, 하, 좌, 우로 붙어 있는 경우 서로 연결되어 있는 것으로 간주
- 얼음 틀의 모양이 주어졌을 때 생성되는 총 아이스크림의 개수를 구하는 프로그램을 작성하라.

- 입력 조건
    - 첫째 줄: 1 <= N, M <= 1000
    - 둘째 줄 ~ N+1째 줄까지 얼음 틀의 형태
    - 구멍이 뚫려있으면 0, 그렇지 않으면 1
- 출력 조건
    - 한 번에 만들 수 있는 아이스크림의 개수

- 입력 예시
    ```python
    """
        00110
        00011
        11111
        00000
    """
    ```
- 출력 예시: 3

````python
"""
    4 5
    00110
    00011
    11111
    00000
"""
# ❌ 풀지 못한 문제
r, c = map(int, input().split()) # n, m 대신 r, c로 변경
ice_tool = []
for _ in range(r):
    ice_tool.append(list(map(int, input())))

# 1. 특정 지점의 주변 상, 하, 좌, 우를 살펴본 뒤 주변 지점 중에서 값이 '0'이면서 방문하지 않은 지점이 있다면 해당 지점을 방문한다.
# 2. 방문 지점에서 다시 상, 하, 좌, 우를 살펴보면서 방문을 진행하면, 연결된 모든 지점을 방문할 수 있다.
# 3. 1 ~ 2번 과정을 모든 노드에 반복하며 방문하지 않은 지점의 수를 센다.
def dfs(x, y):
    print(f"\n[DFS 호출] 위치: ({x}, {y})")

    if not (0 <= x < r and 0 <= y < c): # 주어진 범위에 밖이라면,
        print(f"  → 범위 밖이라 return False")
        return False
    
    if not (ice_tool[x][y] == 0): # 현재 노드를 이미 방문했다면,
        print(f"  → 이미 방문했거나 벽(1)이라 return False")
        return False
    
    print(f"  → 구멍(0) 발견! 방문 처리: ice_tool[{x}][{y}] = 1로 변경")
    ice_tool[x][y] = 1 # 해당 노드 방문 처리

    # 상, 하, 좌, 우 재귀적으로 호출
    print(f"  → 상({x-1}, {y}) 호출 시도")
    dfs(x-1, y)
    print(f"  → 좌({x}, {y-1}) 호출 시도")
    dfs(x, y-1)
    print(f"  → 하({x+1}, {y}) 호출 시도")
    dfs(x+1, y)
    print(f"  → 우({x}, {y+1}) 호출 시도")
    dfs(x, y+1)

    print(f"  ← ({x}, {y}) 모든 방향 탐색 완료 → True 반환")
    return True
        
    
# 모든 노드 위치를 돌면서 음료수 채우기
result = 0
for i in range(r):
    for j in range(c):
        print(f"\n=== 메인 루프: ({i}, {j}) 검사 ===")
        if dfs(i, j) == True:
            result += 1
            print(f"★★★ 새로운 아이스크림 발견! 현재까지 {result}개 ★★★")

print(f"\n최종 아이스크림 개수: {result}")
````

## 260606 - 실전 - 미로 탈출

- N x M 크기의 직사각하여 미로가 있다.
- 미로에는 여러 괴물을 있어서 피해 탈출해야 한다.
- 주인공의 위치는 (1, 1)이고 미로의 출구는 (N, M)에 위치한다.
- 한 번에 한 칸씩 이동할 수 있다.
- 이 때 괴물이 있는 부분은 0, 없는 부분은 1로 표시되어 있다.
- 미로는 반드시 탈출할 수 있는 형태로 제시된다.
- 주인공이 탈출하기 위해 움직여야 하는 최소 칸의 개수를 구하라.
- 단, 시작 칸과 마지막 칸을 모두 포함하여 계산한다.

- 입력 조건
    - 첫째 줄: 4 <= N, M <= 200
    - 둘째 줄부터 0과 1로 이루어진 미로의 정보 제공
    - 각 수는 공백 없이 붙어서 입력으로 제시된다.
    - 또한 시작 칸과 마지막 칸은 항상 1이다.
- 출력 조건
    - 칸의 개수를 출력한다.

- 입력 예시
    ```python
    """
        5 6
        101010
        111111
        000001
        111111
        111111
    """
    ```
- 출력 예시: 10

````python
"""
    5 6
    101010
    111111
    000001
    111111
    111111
"""
# ❌ 풀지 못한 문제
from collections import deque

# 최단 거리 문제이므로, BFS 사용
r, c = map(int, input().split())
map_list = [list(map(int, input())) for _ in range(r)]

# ====================================
# 네 방향 선택
move_types = [(1, 0), (-1, 0), (0, 1), (0, -1)]

# bfs
def bfs(x, y):
    # 큐 선언 및 시작점 삽입
    queue = deque([(x, y)])
    
    # 큐가 빌 때까지 반복
    while queue:
        # 첫 번째 큐 제거
        cr, cc = queue.popleft()

        # 네 방향 분석
        for dr, dc in move_types:
            # 다음 방향 선택
            nr, nc = cr + dr, cc + dc

            # 공간을 벗어날 경우 무시
            if not (0 <= nr < r and 0 <= nc < c): continue

            # 괴물이 있을 경우, 무시
            if map_list[nr][nc] == 0: continue

            # 이미 방문한 칸일 경우 무시
            if map_list[nr][nc] > 1: continue

            # count는 큐에 삽입한 횟수를 셀 뿐, 출구까지 경로 길이가 아님
            # BFS로 최단 거리를 구하는 정석은
            # 격자 자체에 거리를 누적하는 방법
            map_list[nr][nc] = map_list[cr][cc] + 1
            queue.append((nr, nc))

    # 카운트 출력
    return map_list[r-1][c-1]

bfs(0, 0)
````
