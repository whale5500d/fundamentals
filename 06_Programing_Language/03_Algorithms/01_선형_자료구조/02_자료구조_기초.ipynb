# 꼭 필요한 자료구조 기초

- 탐색이란 많은 양의 데이터 중 원하는 데이터를 찾는 과정
    - 프로그래밍에서는 그래프, 트리 등의 자료구조 안에서 탐색하는 문제를 자주 다룸
    - 대표적으로 DFS, BFS
        - 이 두 알고리즘의 원리를 제대로 이해해야 탐색 문제 유형을 풀 수 있다.
        - DFS, BFS를 제대로 이해하기 위해 스택과 큐에 대한 이해가 전제되어야 한다.

## 자료구조

- 자료구조란, 데이터를 표현하고 관리하고 처리하기 위한 구조
    - 스택과 큐는 자료구조의 기초 개념
        - 삽입, 삭제라는 두 가지 핵심 함수로 구성된다.
        - 실제로는 그 외 오버플로, 언더플로를 고민해야 함.
            - 오버플로(Overflow)는 특정 자료구조에 **데이터가 가득 찬 상태**에서 **삽입 연산 수행**할 때 발생
            - 언더플로(Underflow)는 특정 자료구조에 **데이터가 없는 상태**에서 **삭제 연산 수행**할 때 발생

### 스택(Stack)
- 선입후출(FILO) 구조 또는 후입선출(LIFO) 구조를 가진 자료구조
- 파이썬에서는 append, pop 메서드를 이용하면 동일하게 동작 가능

### 큐(Queue)
- 선입선출(FIFO) 구조를 가진 자료구조
- 파이썬에서는 collections 모듈에서 제공하는 deque 자료구조를 활용
    - deque는 스택과 큐의 장점을 모두 채택한 방법
        - 데이터를 넣고 빼는 속도가 리스트 자료형에 비해 효율적
        - queue 라이브러리를 사용하는 것보다 편의성 높음
    - 대부분의 코딩 테스트에서 collections 모듈과 같은 기본 라이브러리 사용을 허용함.
    - deque 객체를 리스트 자료형으로 변경하려면 list() 메서드를 사용하면 됨.

### 재귀 함수(Recursive Function)
- 자기 자신을 다시 호출하는 함수
- 프로그래밍 대회에서는 재귀 함수를 스택을 이용한 함수처럼 동작하도록 자동으로 변경해주는 라이브러리를 사용하기도 한다.
- 이는 파이썬 재귀 함수 제한을 처리하기 위한 또 다른 테크닉이며, 일반적인 코딩 테스트에서는 사용할 일이 적기 때문에 다루지 않는다.
- 수학 시간에 나오는 프랙털(Fractal) 구조와 흡사. 시에르핀스키의 삼각형(Sierpinski Triangle)
- 재귀 함수는 종료 조건이 필요하다.
    - 초반에 등장하는 if 문이 종료 조건 역할을 수행
- 컴퓨터 내부에서는 재귀 함수의 수행은 스택 자료구조를 이용
    - 함수를 계속 호출하고 가장 마지막에 호출한 함수가 수행이 끝나면서 그 앞의 호출 함수가 종료되기 때문
    - 컴퓨터 구조 측면에서 보자면 연속해서 호출되는 함수는 메인 메모리의 스택 공간에 적재되므로 특히 스택 자료구조와 같다.
    - 스택 자료구조를 사용하는 상당수 알고리즘은 재귀 함수를 이용해 간편하게 구현 가능
        - 대표적인 예시: DFS

- 반복문 대신 재귀 함수를 사용했을 때 얻을 수 있는 장점
    - 코드의 간결함. 수학에서 점화식(재귀식)을 그대로 소스코드로 옮겼기 때문
        - 수학에서 점화식은 특정한 함수를 자신보다 더 작은 변수에 대한 함수와의 관계로 표현한 것을 의미
        - 이 개념은 추후 '다이나믹 프로그래밍'으로 이어짐
