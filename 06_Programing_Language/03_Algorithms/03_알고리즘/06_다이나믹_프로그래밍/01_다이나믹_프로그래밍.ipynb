# 다이나믹 프로그래밍

## 중복되는 연산을 줄이자

- 컴퓨터를 활용해도 해결하기 어려운 문제는 무엇인가
    - 최적의 해를 구하기에 시간이 많이 필요하거나
    - 메모리 공간이 매우 많이 필요한 문제 등
- 컴퓨터는 연산 속도에 한계가 있고, 메모리 공간을 사용할 수 있는 데이터의 개수도 한정적이라는 점이 많은 제약을 발생시킨다.
- 그래서 우리는 연산 속도와 메모리 공간을 최대한 활용할 수 있는 효율적인 알고리즘을 작성해야 한다.

- 다만, 어떤 문제는 메모리 공간을 약간 더 사용하면 연산 속도를 비약적으로 증가시킬 수 있는 방법이 있다.
- 대표적인 방법이 다이나믹 프로그래밍 기법으로 "동적 계획법"이라고 표현
- 다이나믹 프로그래밍의 기본적인 아이디어를 소개하고, 다이나믹 프로그래밍의 2가지 방식(탑다운, 바텀업)을 설명한다.
- 특히 다이나믹 프로그래밍을 위해 자주 사용되는 메모이제이션 기법까지 소개한다.

> 프로그래밍에서 다이나믹은 "프로그램이 실행되는 도중에"라는 의미를 갖는다. 하지만 다이나믹 프로그래밍과 동적 할당의 다이나믹은 같은 의미가 아니다. 하지만 다이나믹 프로그래밍에서 다이나믹은 이런 의미가 아니라는 정도만 기억하자. 자료구조에서 동적 할당(Dynamic Allocation)은 프로그램 실행 중에서 프로그램 실행에 필요한 메모리 할당 기법을 의미한다.

- 기존에 어려웠던 문제부터 알아보자.
- 피보나치 수열은 이전 두 항의 합을 현재의 항으로 설정하는 특징이 있는 수열이다.
- 수학자는 점화식을 사용해 수열의 항이 이어지는 형태를 간결하게 표현한다. (점화식: 인접한 항으로 관계식을 표현하는 방법)

$$
a_n = a_{n-1} + a_{n-2}, a_1 = 1, a_2 = 1
$$

- 프로그래밍에서는 이러한 수열을 배열 또는 리스트로 표현 가능
- 수열 자체가 여러 개의 수가 규칙에 따라서 배열된 형태를 의미하기 때문
    - 파이썬에서는 리스트 자료형으로 이를 처리하고 (리스트 자료형이 연결 리스트의 기능을 기본적으로 포함하고 있음)
    - C/C++와 java에서는 배열을 이용해 이를 처리한다.
- 리스트나 배열 모두 "연속된 많은 데이터"를 처리한다는 점이 동일하다.

```python
# 재귀함수로 구현한 피보나치 수열
def fibonacci(x):
    if x == 1 or x == 2:
        return 1
    return fibonacci(x-1) + fibonacci(x-2)

print(fibonacci(4))
```

### 재귀함수로 구현 시 한계점

- 바로 f(n) 함수에 n이 커지면 수행 시간이 기하급수적으로 늘어난다.
- 이 피보나치 수열의 시간 복잡도는 세타 표기법으로 $\Theta$($1.618^{...N}$), 빅오 표기법으로 $O$($2^N$)의 지수 시간이 소요
- n이 커질수록 반복해서 호출하는 수가 많아진다.
- 이런 문제는 다이나믹 프로그래밍을 사용하면 효율적으로 계산할 수 있다.
- 다만, 다이나믹 프로그램을 다음 조건을 만족할 때 사용 가능하다.
    1. ⭐️ 큰 문제를 작은 문제로 나눌 수 있다.
    2. ⭐️ 작은 문제에서 구한 정답은 그것을 포함하는 큰 문제에서도 동일하다.
- 피보나이 수열은 해당 조건을 만족하는 대표적인 문제다.
- 메모이제이션(Memoization) 기법을 사용해 해결 가능하다.

- 메모이제이션은 한 번 구한 결과를 메모리 공간에 메모하고 같은 식을 다시 호출하면 메모 결과를 가져오는 기법이다.
- 메모이제이션은 값을 저장하는 방법이므로 "캐싱(cache)"이라고도 한다.
- 정리하자면 다이나믹 프로그래밍은 같은 문제라면 한 번씩만 풀어서 문제를 효율적으로 해결한다.
- 메모이제이션을 사용하면 시간 복잡도는 O(N)이다.

- 퀵 정렬에서도 소개된 방법이다.
    - 퀵 정렬은 정렬을 수행할 때 정렬할 리스트를 분할하며 전체적으로 정렬될 수 있도록 한다.
    - 이는 분할 정복(divide and conquer) 알고리즘으로 분류된다.
    - 차이점
        - 퀵 정렬: 한 번 원소(Pivot)가 자리를 변경해서 자리를 잡으면, 그 기준 원소의 위치는 더 이상 바뀌지 않는다.
        - 다이나믹 프로그래밍: 한 번 해결했던 문제를 다시 해결한다.

- 메모이제이션과 같이 큰 문제를 해결하기 위해 작은 문제를 호출하는 방식을 탑다운 방식이라 한다.
- 단순 반복문을 이용해 작은 문제부터 큰 문제까지 차근차근 답을 도출하는 방식을 바텀업 방식이라 한다.

```python
# 메모이제이션으로 구현한 피보나치 수열
d = [0] * 100

def fibonacci(x):
    if x == 1 or x == 2:
        return 1
    elif d[x] != 0:
        return d[x]
    else:
        d[x] = fibonacci(x-1) + fibonacci(x-2)
        return d[x]
    
print(fibonacci(99))
```

```python
# 반복문으로 구현한 피보나치 수열(바텀업)
d = [0] * 100

d[1] = 1
d[2] = 1
n = 99

for i in range(3, n+1):
    d[i] = d[i-1] + d[i-2]

print(d[n])
```

- 다이나믹 프로그래밍은 전형적인 바텀업 방식이다.
- 바텀업 방식에서 사용되는 결과 저장용 리스트는 'DP 테이블'이라고 부른다.
- 메모이제이션은 탑다운 방식에 국한되어 사용되는 표현이다.

- 풀이 시간: 20분
- 시간 제한: 1초

- 정수 X가 주어질 때 정수 X에 사용할 수 있는 연산은 아래 4가지
    - X가 5로 나누어 떨어지면, 5로 나눈다.
    - X가 3으로 나누어 떨어지면, 3으로 나눈다.
    - X가 2로 나누어 떨어지면, 2로 나눈다.
    - X에서 1을 뺀다.

- 정수 X가 주어졌을 때, 연산 4개를 적절히 사용하여 1을 만들려고 한다. 연산을 사용하는 횟수의 최솟값을 출력하라.

- 입력 조건
    - 첫째 줄: 2 <= 정수 X <= 30,000

```python
"""
    26
"""
x = int(input())

n = 30000
depths = [0] * (n+1)

# 나의 풀이 - 틀림
def make_one(num):
    for i in range(2, num+1):
        depths[i] = depths[i-1] + 1

        if i % 2 == 0:
            depths[i] = min(depths[i], depths[i // 2] + 1)
        if i % 3 == 0:
            depths[i] = min(depths[i], depths[i // 3] + 1)
        if i % 5 == 0:
            depths[i] = min(depths[i], depths[i // 5] + 1)

    return depths[num]
    

print(make_one(x))

# 답안지
# n = 30000
# d = [0] * (n+1)

# for i in range(2, x+1):
#     d[i] = d[i-1] + 1

#     if i % 2 == 0:
#         d[i] = min(d[i], d[i // 2]+1)
#     elif i % 3 == 0:
#         d[i] = min(d[i], d[i // 3]+1)
#     elif i % 5 == 0:
#         d[i] = min(d[i], d[i // 5]+1)

# print(d[x])
```

## 실전 - 개미 전사

- 풀이 시간: 30분

- 개미 전사는 식량창고를 공격하려고 한다.
- 식량창고는 일직선으로 이어져 있다.
- 각 식량창고에는 정해진 수의 식량을 저장하고 있다.
- 개미 전사는 식량창고를 선택적으로 약탈하여 식량을 빼앗을 예정이다.
- 매뚜기 정찰병 때문에 서로 인접한 식량창고로 가면 안된다.
- 따라서 최소한 한 칸 이상 떨어진 식량창고를 약탈해야 한다.

- 예를 들어 식량 창고 4개가 (1, 3, 1, 5)와 같이 존재한다고 했을 때, 개미 전사는 두 번째 식량창고와 네 번째 식량창고를 선택할 경우 최대값인 8개의 식량을 얻을 수 있다.
- 얻을 수 있는 식량의 최대값을 구하라.

- 입력 조건
    - 첫째 줄: 3 <= 식량 개수 N <= 100
    - 둘째 줄: 공백으로 구분된 저장된 식량의 개수 K (0 <= K <= 1,000)

```python
"""
    4
    1 3 1 5
"""
n = int(input())
array = list(map(int, input().split()))

# 정리
# 1. 두 번째와 네 번째가 가장 많은 값인지 어떻게 알지?
# 순서대로 돌면서 비교 -> 최대값은 구할 수 있음. 차대값은?
# 현재 값들을 어딘가에 저장해두면 좋을 듯
# 생각해보니 두 개일 필요가 없네.
# 2. 가장 많이 얻고 싶으니, 홀수 번째 끼리 합산, 짝수 번째 끼리 합산해서 둘 중 큰 거 반환하자.
# (이렇게하면 결국 모든 창고를 들어가는 방법이라 주제에 맞지 않긴 함.)

# 나의 풀이
# sum_odd = 0
# sum_even = 0

# for index, item in enumerate(array):
#     if index % 2 == 0:
#         sum_even += item
#     else:
#         sum_odd += item

# print(max(sum_odd, sum_even))

# 답안지
d = [0] * 100

d[0] = array[0]
d[1] = max(array[0], array[1])

for i in range(2, n):
    d[i] = max(d[i-1], d[i-2]+array[i])

print(d[n-1])
```

### 문제 해설

- i번째 식량을 털지 말지는 i-1번째, i-2번째에서 결정
    1. i-1 번째 식량창고를 털기로 결정한 경우, i번째 식량창고를 털 수 없다.
    2. i-2 번째 식량창고를 털기로 결정한 경우, i번째 식량창고를 털 수 있다.
    3. 1번, 2번 케이스 중 더 많은 식량을 털 수 있는 경우를 선택한다.
- i-3 번째는 고려할 필요가 없다.
- 따라서 점화식은 아래와 같다.

$$
a_i = max(a_{i-1}, a_{i-2} + k_i)
$$
- $k_i$는 i번째 식량창고에 있는 식량의 양

### 문제 해설

- 다이나믹 프로그래밍의 기초 예제에서 빠질 수 없는 타일링 문제 유형
- 결과를 어떤 수로 나눈 결과를 출력하는 내용이 있는 경우가 많음.
- 이 문제에서도 796,796으로 나눈 나머지를 출력하라고 하는데, 단지 결과값이 굉장히 커질 수 있기 때문
- 따라서 값을 계산할 때마다 특정한 수로 나눈 나머지만 취하도록 한다.
- (추가 설명: 정답이 1이면, 실제 경우의 수는 1, 796797, ..., (796796 * n) + 1임. 796796이라는 값을 이용해 충분한 간격을 두었으니, 정답이 1이라고 해서 해당 알고리즘이 중복된 값을 반환하는 일이 없을 것임을 문제에서 가정했다는 의미)

- 이 문제도 그림을 그리면 어렵지 않다.
- N=3일 때 경우의 수는 5개이다. (머릿속에 그리기)
    1. 왼쪽부터 i-1까지 길이가 덮개로 채워져 있으면, i에선 2 * 1의 덮개 1개를 넣는 경우밖에 없다.
    2. 왼쪽부터 i-2까지 길이가 덮개로 채워져 있으면, 아래 2가지 경우만 존재한다.
        1. 1 * 2 덮개 2개를 넣는 경우
        2. 2 * 2 덮개 1개를 넣는 경우
        3. (2 * 1 덮개 2개를 넣는 경우는 제외, 앞서 이미 고려했기 때문)

- 이 문제도 i-3 번째 이하의 위치에 대한 최적의 해를 고려할 필요가 없다.
- 왜냐하면 사용할 수 있는 덮개의 형태가 최대 2*2 크기의 형태이기 때문이다.
- 다시 말해 바닥을 채울 수 있는 형태는 위에서 언급한 경우밖에 없다.
- 점화식은 $a_i = a_{i-1} + a_{i-2}*2$로 세울 수 있다.

## ⚠️ 실전 - 효율적인 화폐 구성

- 풀이 시간: 30분
- 시간 제한: 1초
- 메모리 제한: 128MB

- N가지 종류의 화폐가 있다. 이 화폐의 개수를 최소한으로 이용해서 그 가치의 합이 M원이 되도록 만드는 프로그램을 구성하라.
- 이 때 각 화폐는 몇 개라도 사용할 수 있으며, 사용한 화폐의 구성은 같지만 순서만 다른 것은 같은 경우로 구분한다.

- 입력 조건
    - 첫째 줄: 1 <= N <= 100, 1 <= M <= 10,000
    - 둘째 줄부터 N개의 화폐의 가치가 주어진다. 화폐 가치는 10,000보다 작거나 같은 자연수이다.
- 출력 조건
    - 첫째 줄: M원을 만들기 위한 최소한의 화폐 개수를 구한다.
    - 불가능할 경우 -1을 출력한다.

```python
"""
    2 15
    2
    3
"""
n, m = map(int, input().split())
array = []
for i in range(0, n):
    array.append(int(input()))

# 나의 풀이
# 최소한의 개수를 사용해야 하므로, 큰 수 우선 사용
# 큰 수보다 작으면 그 다음 큰 수로 뺄셈하기
# 배열에 모든 수를 사용했는데 값이 남으면, -1 출력하기
# 점화식: a_i와 a_{i-1} 사이의 관계식
# array.sort(reverse=True) # 내림차순 정렬
# count = 0
# while m > 0:
#     for _, cash in enumerate(array):
#         while m > cash:
#             m -= cash
#             count += 1
    
#     if m > 0:
#         break

# print(count)

# 답안지

# 초기화
d = [10001] * (m+1)
d[0] = 0

# 연산 진행
for i in range(n):
    for j in range(array[i], m+1):
        if d[j - array[i]] != 10001:
            d[j] = min(d[j], d[j-array[i]]+1)

if d[m] == 10001:
    print(-1)
else:
    print(d[m])
```

### 문제 해설

- 이 문제는 그리디에서 다루었던 거스름돈 문제와 거의 동일
- 단지 화폐 단위에서 큰 단위가 작은 단위의 배수가 아님.
- 그렇기 때문에 그리디 알고리즘을 사용했던 예시처럼 매번 가장 큰 화폐 단위부터 처리하는 방법으로는 해결 불가하므로 다이나믹 프로그래밍을 사용해야 함.

- 이 문제는 적은 금액부터 큰 금액까지 확인하며 차례대로 만들 수 있는 최소한의 화폐 개수를 찾아야 한다.
- 금액 i를 만들 수 있는 최소한의 화폐 개수를 a_i, 화폐의 단위를 k라고 했을 때 다음과 같은 점화식을 작성할 수 있다. a_{i-k}는 금액 (i-k)를 만들 수 있는 최소한의 화폐 개수를 의미한다.
    - a_{i-k}를 만드는 방법이 존재하는 경우, a_i = min(a_i, a_{i-k} + 1)
    - a_{i-k}를 만드는 방법이 존재하지 않는 경우, a_i = 10_001

- 이 점화식을 모든 화폐 단위에 해당하여 차례대로 적용하면 된다. 실제로 문제를 풀기 위해서는 가장 먼저 K의 크기만큼 리스트를 할당한다. 이후에 각 인덱스를 금액으로 고려하여 메모이제이션을 진행한다. 예를 들어 N=3, K=7이고 각 화폐의 단위가 2, 3, 5인 경우를 생각한다.

1. 초기화
    - 각 인덱스에 입력 불가능한 수를 설정
    - M의 최대 크기가 10_000이므로 불가능한 수 10_001을 초기에 설정한다.
    - 0원의 경우, 화폐를 하나도 사용하지 않았을 때 만들 수 있으므로 값을 0으로 설정

2. 연산
    - 2, 3, 5
    1. 첫 번째 화폐 단위 2 확인
        - 2원짜리 화폐를 1개 사용하여 2원을 만들 수 있음. (`d[2] = 1`, $a_2 = a_0 + 1$)
        - 2원짜리 화폐를 2개 사용하여 4원을 만들 수 있음. (`d[4] = 2`, $a_2 = a_2 + 1$)
        - 점화식 기준으로 리스트 갱신
    2. 두 번째 화폐 단위 3 확인
        - 2원짜리 화폐를 1개, 3원짜리 화폐를 1개 사용하여 5원을 만들 수 있음. ($a_5 = a_2 + 1$, `d[5] = d[2] + d[3]`)
        - 점화식 기준으로 리스트 갱신
    3. 세 번째 화폐 단위 5 확인
        - 2원짜리 화폐를 1개, 5원짜리 화폐 1개 사용하여 7원을 만들 수 있음. ($a_7 = a_2 + 1$, `d[7] = d[2] + d[5]`)
        - 기존에 a_7의 값은 3(3개, 2+2+3)이었으나, 현재 2개만 사용해도 되므로, 더 작은 값으로 갱신 
        - 점화식 기준으로 리스트 갱신

- 결과적으로 7원을 만들기 위한 최소의 화폐 개수는 2개이다.
