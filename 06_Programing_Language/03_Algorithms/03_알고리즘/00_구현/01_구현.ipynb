# 아이디어를 코드로 바꾸는 구현

## 피지컬로 승부하기

- 코딩 테스트에서 구현(Implementation)이란 머릿속에 있는 알고리즘을 코드로 구현하는 것이다.
- 구현 문제 유형은 모든 범위의 문제 유형을 포함하는 개념이므로, 대부분 별도의 유형으로 다루지 않는다.
- 하지만 취업을 목표로 하는 경우, 구현이 중심이 되는 문제가 자주 출제되기에 다루고자 한다.
- 이를 위해 프로그래밍 언어의 문법을 정확히 알아야 하고, 요구사항을 모두 반영해야 하고, 실수 없어야 한다.
- 흔히 구현 유형은 '풀이를 떠올리기 쉽지만 소스코드로 옮기기 어려운 문제'를 의미한다.

- 실제 ACM-ICPC, Google Code Jam 등의 대회에 자주 참가하는 사람들이 구현 유형의 문제를 보면
- '알고리즘은 설계했는데 구현이 먼저 풀 수 있는 문제가 없을 때 푸는 게 좋다'고 설명한다.

- 구현하기 어려운 문제 예시
    - 알고리즘은 간단한데 코드가 지나칠 만큼 길어지는 문제
    - 특정 소수점 자리까지 출력해야 하는 문제
    - 문자열이 입력으로 주어졌을 때 한 문자 단위로 끊어서 리스트에 넣어야 하는(파싱해야 하는) 문제 등
    - (소결) 대체로 사소한 조건 설정이 많은 문제일수록 구현이 까다롭다.
    - (소결) 프로그래밍 문법을 정확하게 숙지하지 못했거나, 라이브러리 사용 경험이 부족하면 불리하다.

- 완전 탐색, 시뮬레이션 유형을 모두 '구현' 유형으로 묶어서 다루고 있다.
    - 완전 탐색: 모든 경우의 수를 주저 없이 다 계산하는 방법
    - 시뮬레이션: 문제에서 제시한 알고리즘을 한 단계씩 차례대로 직접 수행하는 방법
    - 둘 다 구현이 핵심이 되는 경우가 많기 때문이다.

## 구현 시 고려해야 할 메모리 제약 사항

### C/C++에서 변수의 표현 범위
- 전통적으로 정수형(Integer)를 표현할 때는 int 자료형을 주로 사용하며 이 자료형의 크기는 4바이트다.
- 특히 C, C++, Java 등을 이용할 떄 int 자료형을 필수적으로 이용한다.
- 기본 int 자료형의 표현 범위는 `-2_147_483_648` ~ `2_147_483_647`
    - 이 말은 int 자료형으로 처리하면 이 밖의 범위의 수는 처리할 수 없다는 의미이다.
- 더 큰 수를 처리하려면 8바이트인 long long과 같은 자료형을 사용하는데, 이 또한 `-9_223_372_036_854_775_808` ~ `9_223_372_036_854_775_807`보다 큰 수를 처리할 수 없다.
- 따라서 훨씬 큰 수를 담을 변수를 만들려면 흔히 BigInteger 클래스를 구현하거나 이용해야 한다.

- 자바의 경우 BigInteger를 표준 라이브러리로 지원하지만, C++의 경우 포함되어 있지 않다.
- 그렇다고 코딩 테스트 중에 직접 작성하기는 어렵다.
- 보통 외부 라이브러리 형태 그래도 가져와 사용한다.
- 하지만 이건 검색과 외부 라이브러리 사용 가능한 코딩 테스트 환경일 때일 뿐이다.
- 대체로 long long에서 다룰 수 있는 수보다 더 큰 정수를 처리하는 문제는 출제되지 않는다.

- 반면 파이썬에서는 직접 자료형을 지정할 필요가 없으며, 매우 큰 수의 연산 또한 기본으로 지원한다.
- 다만, 실수형 변수는 다른 언어와 마찬가지로 유효숫자에 따라서 연산 결과가 원하는 값이 나오지 않을 수 있다.

### 파이썬의 리스트 크기
- 파이썬에서 여러 개의 변수를 사용할 땐 리스트를 이용한다.
- 파이썬에서 리스트를 이용할 때 코딩 테스트의 메모리 제한을 고려해야 한다.
- 대체로 코딩 테스트에서는 128 ~ 512MB로 메모리를 제한한다.
- 알고리즘 문제 중 때로는 수백만 개 이상의 데이터를 처리해야 하는 문제가 출제되곤 한다.
- 이럴 때는 메모리 제한을 염두하고 코딩해야 한다.

- 앞서 다룬 int 자료형의 데이터 개수에 따른 메모리 사용량을 확인해보자.
    - 데이터의 개수(리스트의 길이) = 1_000일 경우, 메모리 사용량 = 4KB
    - 데이터의 개수(리스트의 길이) = 1_000_000일 경우, 메모리 사용량 = 4MB
    - 데이터의 개수(리스트의 길이) = 10_000_000일 경우, 메모리 사용량 = 40MB
- 파이썬에서는 별도의 자료형을 명시할 필요가 없지만, 시스템 내부적으로는 메모리를 차지한다.
- 따라서 구현상 복잡함이 적은 편이지만 데이터 처리량이 많을 때는 꼭 메모리 제한을 고려해야 한다.
    - 선언된 리스트 중 1_000만 이상인 리스트가 있다면 용량 제한으로 풀 수 없을 수 있다는 점을 기억한다.

- 다만, 이런 문제 또한 드물다. 너무 많은 시간이 소요되며 채점 환경에서 문제가 발생할 수 있기 때문이다.
    - 입출력 속도는 프로그래밍 언어마다 조금씩 다르며, 일종의 테크닉이 필요에 따라 사용된다.
    - 모든 언어의 입출력 속도까지 고려해 시간 제한을 설정하는 것은 출제자 입장에서도 번거롭기 때문이다.

- (소결) 일반적인 수준에서는 메모리 사용량 제한보다 적은 크기의 메모리를 사용해야 한다는 점 정도 기억한다.

## 채점 환경
- 실제 온라인에서 사용되는 채점 환경의 시스템 제한은 기관마다, 문제마다 다르다.
- 출제자가 빠른 동작하는 프로그램을 원한다면 시간 제한은 1초와 같이 짧을 것이다.
- 일반적인 환경에서는 채점 시스템의 시간 제한 및 메모리 제한 정보가 적혀 있다.

- 파이썬은 C/C++에 비해 동작 속도가 느리다.
- 그래서 파이썬을 선택했을 때 C/C++에 비해 2배의 수행 시간 제한을 적용하기도 한다.
- 일반적인 환경에서는 파이썬으로 제출한 코드가 1초에 2_000만 번 연산을 수행한다고 가정하면 무리없다.

- 제한 시간이 1초이고, 데이터가 100만 개라면, 보통 시간복잡도 O(NlogN)이내의 알고리즘을 이용해야 한다.
    - N = `1_000_000`일 때, $Nlog_2N$은 약 `20_000_000`이기 때문이다.

- 따라서 시간 제한과 데이터의 개수를 확인한 후, 시간복잡도가 어느 정도 필요한지 예측할 수 있어야 한다.

## 구현 문제 접근 방법
- 보통 구현 문제는 사소한 입력 조건 등을 문제에 명시해주어 문제 길이가 꽤 긴 편이다.
- 고차원적인 사고력을 요구하지 않는 편이라 문법에 익숙하다면 쉽게 풀 수 있다.
- 구현 문제는 문자열을 처리하거나 큰 정수를 처리하는 문제가 많은데,
    - C/C++나 자바에서는 문자열 처리가 까다롭고, 큰 정수를 처리하는 라이브러리를 별도로 사용해야 한다.
    - 반면 파이썬은 기본 문법만 알아도 상대적으로 쉽게 해결할 수 있다.

- C/C++ VS 파이썬
    - 실무에서 파이썬으로 프로그램을 개발할 때는 GPU를 연동한다.
    - 반복적인 행렬 계산을 요구하는 복잡한 수학 문제를 풀 때는 C로 작성된 파이썬 코어 소프트웨어가 동작한다.
    - 그렇기 때문에 파이썬을 쓴다고 해서 항상 동작 속도가 느린 것은 아니다.
    - 하지만 알고리즘 코딩 테스트 환경에서는 GPU 연산을 쓰는 경우가 없다.
    - 파이썬(Pypy3)
        - 자동 채점 방식을 이용하는 환경에서는 Pypy3를 지원하는 곳이 늘고 있다.
        - Pypy3는 파이썬3 문법을 지원하며, 실행 속도가 파이썬3보다 빠르다.
        - 이 말은 코딩 테스트에서 Pypy3를 선택한다면 실행 시간을 줄일 수 있다는 의미다.
        - 삼성전자 공채에서는 Pypy3를 채점에 이용한다.
        - 특히 반복문이 많을수록 Pypy3와 파이썬3의 속도가 차이 난다.
        - Pypy3는 때론 C/C++와 견줄 만큼 빠르다. 대략 1초에 2,000만 번에서 1억 번 연산할 수 있다.
        - 만약 코딩 테스트 환경이 Pypy3도 지원한다면 이를 이용하도록 하자.
    - C/C++
        - C/C++를 이용한다고 해서 구현이 무조건 어려운 것은 아니다.
        - C/C++를 주 언어로 사용하는 숙련도 높은 랭커들은 자신만의 코드 노트가 있어 이를 활용한다.
        - ACM-ICPC와 같은 국제 대회에서는 구현인 복잡한 경우, 팀 노트(A4, PDF)를 준비할 수 있다.
        - 이럴 때 자신만의 라이브러리가 이미 구축되어 있다면 충분히 난이도를 낮출 수 있다.
        - 하지만 일반적으로 C/C++는 고려해야 할 사항이 많아 초보자에게는 어렵다.

- API 개발 문제 또한 구현 유형과 상당히 맞닿아 있다.
    - 예시: 카카오 공채 때 API 개발 문제가 출제된 적이 있다.
    - 이때 카카오 문제 풀이 서버와 통신하는 프로그램 모듈을 작성해야 했다.
    - 이는 알고리즘 문제와 별개로 웹 서버, 데이터 분석에 대한 기초 지식도 필요하다.
    - 이때, C++나 자바에 비해 파이썬은 간결하고 직관적인 코드 라이브러리를 사용할 수 있다.
    - 파이썬을 사용한다면 API 개발 문제를 보더라도 상대적으로 무난하게 대처할 수 있다.

## 예시 - 상하좌우
- 여행가 A는 N x N 크기의 정사각형 공간 위에 서 있다.
- 이 공간은 1 x 1 크기의 정사각형으로 나누어져 있다.
- 가장 왼쪽 위 좌표는 (1, 1)이며, 가장 오른쪽 아래 좌표는 (N, N)에 해당된다.
- A는 상, 하, 좌, 우 방향으로 이동할 수 있으며, 시작 좌표는 항상 (1, 1)이다.
- 계획서에는 한 줄에 띄어쓰기를 기준으로 L, R, U, D 중 하나의 문자가 반복적으로 적혀있다.
- (L: 왼쪽으로 한 칸 이동, R: 오른쪽, U: 위로, D: 아래로)
- A가 N x N 크기의 정사각형을 벗어나는 움직임은 무시된다.

- 예시
    - 입력: N = 5, 계획서 = `R R R U D D`
    - 출력: 3 4

- 계획서와 N의 크기가 주어졌을 때 A가 최종적으로 도착할 지점의 좌표를 출력하는 프로그램을 완성하라.

```python
"""
    5
    R R R U D D
"""
n = int(input())
plan_list = list(map(str, input().split()))

position = [1, 1]

# 방법 1. 나의 방법
# for step in plan_list:
#     if step == 'R':
#         if position[1] == n: continue
#         position[1] += 1
#     elif step == 'L':
#         if position[1] == 1: continue
#         position[1] -= 1
#     elif step == 'U':
#         if position[0] == 1: continue
#         position[0] -= 1
#     elif step == 'D':
#         if position[0] == n: continue
#         position[0] += 1

# 방법 2. 해설
d_row = [0, 0, -1, 1] # L R U D 순
d_col = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']

for step in plan_list:
    # 우선 이동하기
    nr, nc = position[0], position[1]
    for i in range(len(move_types)):
        if step == move_types[i]:
            nr = position[0] + d_row[i]
            nc = position[1] + d_col[i]
            break

    # 이동한 결과가 범위를 벗어날 경우, 무시
    if nc < 1 or nr < 1 or nc > n or nr > n:
        continue

    # 벗어나지 않았을 경우, 이동 수행
    position[0] = nr
    position[1] = nc

print(f"({position[0]}, {position[1]})")
```

### 문제 해설
- 요구사항대로 구현하면 연산 횟수는 이동 횟수에 비례한다.
- 따라서 이동 횟수가 N인 경우 시간 복잡도는 O(N)이다.
- 시간복잡도는 넉넉한 편이다.
- 이런 문제는 일련의 명령에 따라서 계체를 차례대로 이동한다는 점에서 시뮬레이션 유형으로 분류된다.
- 논리적 사고력을 확인할 수 있는 가장 기본 난이도로 적합한 문제이다.
