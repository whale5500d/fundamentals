# 기준에 따라 데이터를 정렬하기

## 정렬 알고리즘 개요

- 데이터를 특정한 기준에 따라서 순서대로 나열하는 것
- 정렬을 배우면 다음에 배울 이진 탐색(Binary Search)이 가능
- 정렬 알고리즘은 이진 탐색의 전처리 과정이기 때문
- 대표적인 정렬 4가지
    - 선택 정렬
    - 삽입 정렬
    - 퀵 정렬
    - 계수 정렬
- 더불어 파이썬에서 제공하느 기본 정렬 라이브러리를 적용하여 효과적인 정렬 수행 방법도 다룰 예정

- 예제는 모두 오름차순 정렬을 수행한다고 가정
- 내림차순 정렬은 오름차순의 크기 비교를 반대로 수행하면 된다.
- 또한 오름차순 정렬을 수행한 뒤 뒤집기(Reverse)로 내림차순을 만들 수도 있다.
    - 뒤집기 연산은 O(N)의 복잡도를 가지므로 간단히 수행 가능

## 선택 정렬

- 가장 원시적인 방법으로 매번 가장 작은 것을 선택하는 알고리즘
- 가장 작은 데이터를 선택해 맨 앞에 있는 데이터와 바꾸고, 그 다음 작은 데이터를 선택해 앞에서 두 번째 데이터와 바꾸는 과정의 반복

- 스와프란 특정한 리스트가 주어졌을 때 두 변수의 위치를 변경하는 작업
- 다른 프로그래밍 언어에서는 명시적으로 임시 저장용 변수를 사용해 두 원소의 값을 변경해야 한다.


````python
# 선택 정렬 예시
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(len(array)):
    min_index = i
    for j in range(i+1, len(array)):
        if array[min_index] > array[j]:
            min_index = j
    array[i], array[min_index] = array[min_index], array[i] # 스와프

print(array)
````

### 선택 정렬의 시간 복잡도

- 매번 가장 작은 수를 찾기 위해 비교 연산을 수행한다.
- 연산 횟수는 N + (N-1) + ... + 2 (근사치는 N * (N+1) / 2)
- 빅오 표기법으로 O(N^2)로 표현
    - 데이터의 개수가 100배 늘어나면, 수행 시간은 10,000배로 늘어남

- 선택 정렬은 다른 정렬 알고리즘과 비교했을 때 매우 비효율적이다.
- 다만, 특정한 리스트에서 가장 작은 데이터를 찾는 일이 알고리즘 테스트에서 잦으므로 선택 정렬 코드 형태에 익숙해질 필요가 있다.

## 삽입 정렬

- 데이터를 하나씩 확인하며, 각 데이터를 적절한 위치에 삽입하는 정렬 알고리즘
- 선택 정렬에 비해 구현 난이도는 높지만, 실행 시간 측면에서 더 효율적이다.
- 필요할 때만 위치를 바꾸므로 '데이터가 거의 정렬되어 있을 때' 훨씬 효율적이다.
- 특정한 데이터가 적절한 위치에 들어가기 이전에, 그 앞 데이터는 이미 정렬되어 있다고 가정
- 정렬되어 있는 데이터 리스트에서 적절한 위치를 찾은 뒤에 그 위치에 삽입
- 삽입 정렬은 두 번째 데이터부터 시작.
    - 첫 번째 데이터는 정렬되어 있다고 가정

- 삽입 정렬은 정렬이 이루어진 원소가 항상 정렬(오름차순)을 유지하고 있다는 점이 특징
    - 이러한 특징으로 삽입 정렬에서는 특정 데이터가 삽입될 위치를 선정할 때, 삽입될 데이터보다 작으면 그 위치에서 멈추면 된다.

````python
# 삽입 정렬
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(1, len(array)):
    print(f"i: {i}")
    for j in range(i, 0, -1):
        print(f"j: {j}")
        if array[j] < array[j-1]:
            print("before: ", array[j-1], array[j])
            array[j], array[j-1] = array[j-1], array[j]
            print("after: ", array[j-1], array[j])
        else:
            print("break time: ", array[j-1], array[j])
            break

print(array)
````

### 삽입 정렬의 시간 복잡도

- O(N^2)
- 수행 시간을 테스트해보면 선택 정렬과 흡산한 시간이 소요되는 것을 알 수 있다.
- 꼭 기억해야 할 내용은 현재 리스트의 데이터가 거의 정렬되어 있는 상태라면 매우 빠르게 동작한다는 점이다.
- 최선의 경우 O(N)의 시간 복잡도를 가진다.
- 정렬이 거의 되어있는 상태라면 퀵 정렬 알고리즘보다 더 강력하다.

## 퀵 정렬

- 기준 데이터를 설정하고 그 기준보다 큰 데이터와 작은 데이터의 위치를 바꾸는 알고리즘
- 가장 많이 사용되는 알고리즘
- 퀵 정렬과 비교할 만큼 빠른 알고리즘으로 병합 정렬 알고리즘이 있다.
- 이 두 알고리즘은 대부분의 프로그래밍 언어에서 정렬 라이브러리의 근간이 되는 알고리즘이기도 하다.
- 고급 정렬 기법: 병합 정렬, 힙 정렬
- 교환하기 위한 기준을 피벗(Pivot)이라 표현
- 퀵 정렬을 수행하기 전에 피벗을 어떻게 설정할 것인지 미리 명시해야 한다.
- 피벗 설정하고 리스트 분할 방법에 따라서 퀵 정렬 구분

- 대표적으로 호어 분할(Hoare Partition) 방식 기준으로 퀵 정렬 설명
    - 리스트의 첫 번째 데이터를 피벗으로 정한다.
    - 왼쪽부터 피벗보다 큰 데이터를 찾고,
    - 오른쪽부터 피벗보다 작은 데이터를 찾는다.
    - 그다음 큰 데이터와 작은 데이터의 위치를 서로 교환한다.
    - 두 값이 엇갈린 경우, 작은 데이터와 피벗의 위치를 서로 변경한다.
    - 피벗을 기준으로, 왼쪽 데이터끼리 퀵 정렬, 오른쪽 데이터끼리 퀵 정렬 반복
    - 현재 리스트의 데이터 개수가 1개인 경우, 이미 정렬되어 있다고 간주하여 종료한다.

````python
# 퀵 정렬
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array, start, end):
    # 5. 종료 조건
    if start >= end:
        return
    
    # 1. 피벗 설정
    pivot = start
    
    # 2. left, right이 엇갈릴 때까지 변경 반복
    left = start + 1
    right = end
    while left <= right:
        # 3. 변경 과정
        # 3.1. 피벗보다 작은 데이터 찾을 때까지 left 반복
        while left <= end and array[left] <= array[pivot]: left += 1

        # 3.2. 피벗보다 큰 데이터 찾을 때까지 right 반복
        while right > start and array[right] >= array[pivot]: right -= 1

        # 3.3. 엇갈렸다면 left와 피벗 교체
        if left > right:
            array[right], array[pivot] = array[pivot], array[right]

        # 3.4. 엇갈리지 않았다면 left와 right 교체
        else:
            array[left], array[right] = array[right], array[left]

    # 4. left과 right에서 각각 퀵 정렬 수행
    quick_sort(array, start, right-1)
    quick_sort(array, right+1, end)


quick_sort(array, 0, len(array)-1)
print(array)
````

````python
# 파이썬의 장점을 살린 퀵 정렬
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    # 종료 조건
    if len(array) <= 1:
        return array
    
    # 1. 피벗 설정
    pivot = array[0]
    tail = array[1:]
    
    # 2. left, right array 분할 생성
    left_array = [x for x in tail if x <= pivot]
    right_array = [x for x in tail if x > pivot]

    # 분할 이후 왼쪽, 오른쪽 정렬 퀵 정렬 재수행, 전체 리스트 반환
    return quick_sort(left_array) + [pivot] + quick_sort(right_array)

print(quick_sort(array))
````

## 퀵 정렬의 시간 복잡도

- 평균 시간 복잡도는 O(NlogN)
- 시간 복잡도 증명은 초보자가 다루기 어려움. 따라서 직관적인 이해를 돕는 방향으로 설명
    1. 최선의 경우
        - (가정) 피벗값의 위치가 변경되어 분할이 일어날 때마다 정확히 왼쪽 리스트와 오른쪽 리스트를 절반씩 분할한다는 가정
        - (가정) 데이터의 개수가 8개이고, 정확히 절반식 나눈다고 가정
        - 데이터의 개수가 8개이므로, 높이는 log8 = 3 (=컴퓨터공학에서 지수 2 생략)
        - 데이터의 개수 N일 때 높이(깊이 횟수)는 약 logN이라 할 수 있다.
        - 분할이 이루어지는 횟수가 기하급수적으로 감소하게 된다.
        - 데이터의 개수가 많을수록 차이는 매우 극명하게 드러난다.
            - N = 1,000, N*logN은 약 10,000으로 상대적으로 매우 작은 수임을 이해할 수 있다.
            - N = 1,000,000, N*logN은 약 20,000,000
            - (엄밀하지는 않음, 직관적인 결과일 뿐)
        - 퀵 정렬은 앞서 다루었던 선택 정렬, 삽입 정렬에 비해 압도적으로 빠르게 동작하리라 추측할 수 있다.
    2. 최악일 경우
        - O(N^2)도 가능
        - 데이터가 무작위로 입력되는 경우, 퀵 정렬은 빠르게 동작한다.
        - 예시만큼 리스트의 가장 외쪽 데이터를 피벗으로 삼을 때, '이미 데이터가 정렬되어 있는 경우' 매우 느리게 동작함.
        - 앞서 다룬 삽입 정렬의 경우 이미 데이터가 정렬되어 있을 때 빠르게 동작하는 예시와 반대가 됨.
        - 실제로 C++과 같이 퀵 정렬을 기반으로 작성된 정렬 라이브러리를 제공하는 프로그래밍 언어는 최악의 경우에도 O(NlogN)가 되는 것을 보장할 수 있도록 피벗값을 설정할 때 추가적인 로직을 더해준다.
        - 파이썬 또한 마찬가지로 뒤에 설명할 기본 정렬 라이브러리를 이용하면 O(NlogN)을 보장해준다.
            - 구체적인 로직은 책에서 다루지 않으므로, 추가적인 학습 필요

## 계수 정렬
- 한 줄 요약: 특정한 조건이 부합할 때만 사용할 수 있지만 매우 빠른 정렬 알고리즘

- 예시
    - 모든 데이터가 양의 정수인 상황을 가정
        - 데이터의 개수가 N, 데이터 중 최대값 K일 때,
        - 계수 정렬은 최악의 경우 수행 시간 O(N+K)를 보장

- 조건(가정)
    - '데이터의 크기 범위가 제한되어 정수 형태로 표현할 수 있을 때만' 사용할 수 있다.
    - 데이터의 값이 실수형 데이터(무한한 범위)가 주어지는 경우 계수 정렬은 사용하기 어려움
    - 일반적으로 최소값와 최대값 차이가 1,000,000을 넘지 않을 때 효과적
        - 예시: [0, 100] 범위의 성적 데이터를 정렬할 때 계수 정렬은 효과적

- 방법
    - 별도의 리스트를 선언하고 그 안에 정렬에 대한 정보를 담는 특징
    - 계수 정렬은 앞서 다룬 "직접 비교 방식"을 사용하지 않음.

- 이러한 특징을 갖는 이유
    - 계수 정렬을 이용할 때 모든 범위를 담을 수 있는 크기의 리스트(배열)을 선언해야 하기 때문
    - 최대값과 최소값의 차이가 1,000,000이라면, 총 1,000,001개의 데이터가 들어갈 수 있는 리스트를 초기화해야 함.
        - 0 ~ 1,000,000까지 때문에 총 1,000,001개

### 예시

- 입력 데이터: 7, 5, 9, 0, 3, 1, 6, 2, 9, 1, 4, 8, 0, 5, 2
    - 앞서 다룬 예시와 다르게 많은 데이터가 존재하는 경우
    - 현재의 경우 최대값 9, 최소값 0

1. 먼저 **최대값과 최소값의 범위가 모두 담길 수 있도록** 하나의 리스트를 생성
    - 데이터의 범위가 0~9이므로 크기가 10인 리스트를 선언
2. 데이터를 하나씩 확인하며 데이터의 값과 동일한 인덱스의 데이터를 1씩 증가 (계수 정렬 완료)
    - 결과적으로 리스트에 데이터가 몇 번 등장했는지 횟수 기록됨.
3. 리스트의 첫 번째 데이터부터 하나씩 그 값만큼 인덱스를 출력 (인덱스가 곧 실제 값)
    - 이 리스트에 저장된 데이터 자체가 정렬된 형태 그 자체

````python
# 예시
# (가정) 모든 원소의 값이 0보다 크거나 같다고 가정
array = [7, 5, 9, 0, 3, 1, 6, 2, 9, 1, 4, 8, 0, 5, 2]

# 1. 모든 범위를 포함하는 리스트 선언 (모든 값은 0으로 초기화)
count = [0] * (max(array) + 1)

# 2. 각 데이터에 해당하는 인덱스의 값 증가
for i in range(len(array)):
    count[array[i]] += 1

# 3. count에 기록된 정렬 정보를 돌면서 등장한 횟수만큼 추
for i in range(len(count)):
    for j in range(count[i]):
        print(i, end=' ')
````

## 계수 정렬의 시간 복잡도
- 양의 정수인 상황에서 데이터의 개수를 N, 데이터 중 최대값의 크기가 K라면, 시간 복잡도는 O(N+K)
- 앞에서부터 데이터를 하나씩 확인하면서 리스트에서 적절한 인덱스의 값을 1씩 증가시킬 뿐 아니라(N),
- 추후 리스트의 각 인덱스에 해당하는 값을 확인할 때 데이터 중 최대값의 크기만큼 반복 수행(K)해야 하기 때문
- 따라서 범위만 한정되어 있다면, 효과적으로 사용 가능 + 항상 빠르게 동작
- 현존하는 정렬 알고리즘 중 기수 정렬(Radix Sort)과 더불거 가장 빠름
    - 기수 정렬은 계수 정렬에 비해 동작은 느리지만, 처리 정수의 크기는 더 크다.
    - 기수 정렬은 다만 알고리즘 원리나 소스코드는 더 복잡하다.
    - 다행히 기수 정렬을 이용해야만 해결할 수 있는 문제는 코딩 테스트에 거의 출제되지 않는다.
    - 책에서는 기수 정렬에 대해 자세히 다루지 않는다.

## 계수 정렬의 공간 복잡도
- 공간 복잡도도 O(N+K)
- 상황에 따라 심각한 비효율성 초래
- 예시
    - (가정) 데이터가 0, 999,999 단 2개만 존재한다고 가정
    - 이럴 때도 리스트의 크기가 100만 개가 되도록 선언해야 함
    - 따라서 항상 사용할 수 있는 알고리즘이 아님
    - (적합) 동일한 값을 가지는 데이터가 여러 개 등장할 때 적합
        - 성적의 경우 100점을 맞는 학생이 여러 명일 수 있으므로, 효과적
    - 반면, 퀵 정렬은 일반적인 경우 평균적으로 빠르게 동작하므로,
        - 데이터의 특성을 파악하기 어렵다면 퀵 정렬을 이용하는 것이 효과적
- (결론) 계수 정렬은 크기가 한정되어 있고, 크기가 많이 중복될 수록 효과적이고, 항상 사용할 수는 없다.
    - 조건만 만족한다면 계수 정렬은 정렬해야 하는 데이터의 개수가 매우 많을 때도 효과적
    - 다만, 일반적인 코딩 테스트 환경에서는 메모리 공간상 제약과 입출력 시간 문제로 개수를 1,000만 개 미만으로 출제될 것

## 파이썬의 정렬 라이브러리
- 파이썬은 기본 정렬 라이브러리 sorted() 함수 제공
- sorted()는 퀵 정렬과 동작 방식이 비슷한 병합 정렬로 만들어짐
- 병합 정렬은 일반적으로 퀵 정렬보다 느리지만 최악의 경우에도 시간 복잡도 O(NlogN)을 보장
- 리스트, 딕셔너리 자료형 등을 입력받아서 정렬된 결과를 출력
- 집합 자료형이나 딕셔너리 자료형을 입력받아도 반환되는 결과는 리스트 자료형

````python
# py sorted를 사용하는 방법
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

result = sorted(array)
print(result)
````

````python
# py sort를 사용하는 방법
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

array.sort()
print(array)
````

````python
# py 정렬 라이브러리에서 key를 활용하는 방법
array = [('바나나', 2), ('사과', 5), ('당근', 3)]

def setting(data):
    return data[1]
result = sorted(array, key=setting)
print(result)
````

### 정렬 라이브러리의 시간 복잡도
- 최악의 경우에도 시간 복잡도가 O(NlogN)을 보장
- 우리가 직접 퀵 정렬을 구현할 때보다 더욱 효과적
- 병합 정렬에 기반한다고 했는데, 정확히는 병합 정렬과 삽입 정렬의 아이디어를 더한 하이브리드 방식의 정렬 알고리즘 사용
- 문제에서 별도의 요구가 없다면 단순히 정렬해야 하는 상황에서 기본 정렬 라이브러리를 사용하고,
- 데이터의 범위가 한정되어 있으며 더 빠르게 동작해야 할 때는 계수 정렬을 사용한다.

1. 정렬 라이브러리로 풀 수 있는 문제
    - 단순히 정렬 기법을 알고 있는지 물어보는 문제로, 기본 정렬 라이브러리의 사용 방법을 숙지하고 있으면 쉽게 해결 가능
2. 정렬 알고리즘의 원리에 대해 물어보는 문제
    - 선택 정렬, 삽입 정렬, 퀵 정렬 등의 원리를 알고 있어야 해결 가능
3. 더 빠른 정렬이 필요한 문제
    - 퀵 정렬 기반의 정렬 기법으로는 불가능,
    - 계수 정렬 등 다른 알고리즘을 이용하거나 문제에서 기존에 알려진 알고리즘의 구조적인 개선을 거쳐야 해결 가능

## 260612 - 실전 - 두 배열의 원소 교체

- 풀이 시간: 20분
- 시간 제한: 2초
- 메모리 제한: 128MB
- 기출: 국제 알고리즘 대회

- 두 개의 배열 A와 B이 있다.
- 두 배열은 N개의 원소로 구성되어 있으며, 배열의 원소는 모두 자연수이다.
- 최대 K번의 바꿔치기 연산을 수행할 수 있다.
- 바꿔치기 연산이란 배열 A에 있는 원소 하나와 배열 B에 있는 원소 하나를 골라서 두 원소를 서로 바꾸는 연산을 의미한다.
- 최종 목표는 배열 A의 모든 원소의 합이 최대가 되도록 하는 것이다.
- N, K 그리고 배열 A, B 정보가 주어졌을 때, 최대 K번의 바꿔치기 연산을 수행하여 만들 수 있는 배열 A의 모든 원소의 합의 최대값을 출력하라.

- 입력 조건:
    - 첫째 줄: N, K가 공백으로 구분되어 입력 (1 <= N <= 100,000, 0 <= K <= N)
    - 둘째 줄: 배열 A의 원소들이 공백으로 구분되어 입력 (모든 원소는 10,000,000보다 작은 자연수)
    - 셋째 줄: 배열 B의 원소들이 공백으로 구분되어 입력 (모든 원소는 10,000,000보다 작은 자연수)
- 출력 조건:
    - 최대 K번의 바꿔치기 연산을 수행하여 만들 수 있는 배열 A의 모든 원소의 합의 최대값

- 입력 예시:
    ```python
    """
        5 3
        1 2 5 4 3
        5 5 6 6 5
    """
    ```
- 출력 예시: 26

````python
"""
    5 3
    1 2 5 4 3
    5 5 6 6 5
"""
n, k = map(int, input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))

# 나의 풀이 (구현으로 해결)
# 언제 a의 합이 최대값이 될까?
# b에 존재하는 모든 큰 값이 a에 존재할 때
# a의 최소값을 b의 최대값과 비교했을 때 더 크면 break하고 최대값으로 판단

# 총 시간 복잡도: O(N * K)
# while True: # O(K)
#     min_a = min(a) # O(N)
#     max_b = max(b) # O(N)
    
#     # 종료 조건
#     if k == 0: break
#     if min_a >= max_b: break

#     # 바꿔치기 조건
#     else:
#         min_a_i = a.index(min_a) # O(N)
#         max_b_i = b.index(max_b) # O(N)
#         a[min_a_i], b[max_b_i] = max_b, min_a
#         k -= 1 # k번 만큼 실행

# result = 0
# for value in a:
#     result += value
# print(result)

# 답안지
a.sort() # 오름차순
b.sort(reverse=True) # 내림차순

for i in range(k):
    if a[i] >= b[i]:
        break
    else:
        a[i], b[i] = b[i], a[i]

print(sum(a))
````

### 코드 오답 노트

- 시간복잡도: O(N * K)
- K번의 while 문을 돌면서, 최대(최소)값 구할 때 N번만큼 반복 중
- 매번 최대(최소)값을 구하는 것은 비효율적, 영리하게 최대(최소)값를 구하는 방법 고안 필요

- <u>(아이디어) 한 번 바꿔치기하면 그 다음 큰(작은) 값이 최대(최소)값이 된다.</u>
- 따라서 배열을 오름차순(내림차순)으로 <u>미리</u> 정렬해둔다.

### 문제 해설

- 문제 해결의 기본 아이디어는 매번 배열 A(B)에서 가장 작은(큰) 원소를 골라서, 교체하는 것이다.
- 단, 배열 A에서 가장 작은 원소가 배열 B에서 가장 큰 원소보다 작을 때에만 교체를 수행
- 이러한 과정을 K번 반복하면 된다.

- 따라서 배열 A와 B의 정보가 입력되면, A(B)를 오름차순(내림차순)으로 정렬한다.
- 그리고 두 배열의 첫 번째 인덱스부터 차례대로 비교하면서 A의 원소가 B의 원소보다 작을 때만 교체를 수행한다.

- ⭐️ 이 때 두 배열의 원소가 최대 100,000개까지 입력될 수 있으므로 O(NlogN)을 보장하는 정렬 알고리즘을 이용해야 한다.
