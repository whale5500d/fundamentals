```python
# 데이터 처리에 필요한 10개 항목만 추출하겠습니다. 
# 고객등급(cust_class), 성별(sex_type), 나이(age), 사용서비스수(efct_svc_count), 서비스중지여부 (dt_stop_yn), 미납여부(npay_yn)
# 3개월 평균 요금(r3m_avg_bill_amt), A서비스 3개월 평균요금(r3m_A_avg_arpu_amt), B서비스 3개월 평균요금(r3m_B_avg_arpu_amt), 해지여부(termination_yn)
# 10개의 필요한 컬럼들을 리스트에 넣으면 되고, 결과를 cust 데이터프레임에 저장합니다. 

cust = df[ ["cust_class","sex_type","age","efct_svc_count","dt_stop_yn",
         "npay_yn","r3m_avg_bill_amt","r3m_A_avg_arpu_amt","r3m_B_avg_arpu_amt", "termination_yn"] ]
```

실습의 편의를 위해 컬럼명을 조금 더 간결하고 직관적으로 변경해 보겠습니다.

```python
# 컬럼명을 간결하고 직관적으로 변경해 보겠습니다. 
# rename 함수를 사용하여 컬럼명 변경합니다. 

cust = cust.rename( columns = {"cust_class" : 'class',          # 고객등급
                             "sex_type":'sex',                # 성별
                             "efct_svc_count":'service',      # 사용 서비스수
                             "dt_stop_yn":'stop',             # 서비스 중지 여부
                             "npay_yn":'npay',                # 미납 여부
                             "r3m_avg_bill_amt":'avg_bill',   # 3개월 평균 요금
                             "r3m_A_avg_arpu_amt":"A_bill",   # A서비스 3개월 평균요금
                             "r3m_B_avg_arpu_amt":'B_bill',   # B서비스 3개월 평균요금
                             "termination_yn":'termination' } # 해지 여부
                ) 
```

```python
# cust 데이터프레임의 이름 변경 확인하기
# 컬럼이름이 간결해져서 보기 좋네요.

cust.head()
```

## **2. 결측치 처리하기**
- 결측치 : Missing Value로 데이터에 값이 없는 것을 의미. NA, Null, NaN으로 표시

### <span style="color:RED">[참고]</span> 데이터 변경시 되도록 원본 데이터를 COPY 해서 사용할것!!
파이썬에서 Copy 메소드를 사용하지 않으면 주소값을 복사해서 사용기 때문에 원본 값을 변경 시키게 됩니다.    
따라서 원본 항목을 보전하면서 데이터를 보정하려면 copy 메소드를 사용 해주시기 바랍니다.

```python
# copy() 함수를 이용해서 기존 Cust 데이터프레임을 새로운 cust_fix 데이터프레임으로 복사하기 
cust_fix = cust.copy()
```

```python
# 새로 만든 cust_fix 데이터프레임 확인하기
cust_fix.head()
```

```python
# 이번에는 copy() 함수를 사용하지 않고 직접 대입해서  'test' 데이터프레임 만들어 보겠습니다. 
test = cust
```

```python
# 새로 만든 test 데이터프레임 확인하기
# 내용은 같고 별차이를 모르겠네요. 

test.head()
```

```python
# 뭐가 다른지 알기 위해서, test 데이터프레임의 age 컬럼의 값들을 숫자 0 으로 변경해 보겠습니다.
test['age'] = 0
```

```python
# test 데이터프레임의 age 컬럼의 값들을 확인해 보니 모두 숫자 0으로 나옵니다.
test.head()
```

```python
# 위에서 test 데이터 변경 했으며, Cust 데이터프레임은 변경하지 않은 상태입니다.
# cust 데이터프레임을 확인해 보니, 
# cust 데이터프레임의 age 컬럼의 값들도 숫자 0으로 변경됨을 확인할수 있습니다. 

cust.head()
```

```python
# 앞에서 Copy 함수로 변경한 Cust_fix 데이터 확인 보니
# 예전의 age 컬럼의 값들을 잘 보전하고 있습니다. 

cust_fix.head()
```

```python
# test = cust 처럼 직접 대입하여 사용하면, 주소로 복제되어 
# test 데이터프레임의 변경사항이 cust 데이터프레임에도 똑같이 반영됩니다.
# 그래서, 데이터 변경시에는 사전에 copy() 함수를 활용해서 복제해 두어 
# 변경 이전의 데이터프레임의 값들을 보전하면 좋습니다.
# 전의 cust 데이터프레임을 보관한 cust_fix 데이터프레임을 
# cust 데이터프레임에 다시 복제하도록 하겠습니다.

cust = cust_fix.copy()
```

```python
# cust 데이터프레임의 age 컬럼의 값들이 정상적으로 보입니다. 
cust.head()
```

### <b>결측치 채우기 </b>

사용자가 지정하는 단일값으로 채워서 결측치를 처리하는 방법입니다.

```python
# 결측치 채우기 이전에 먼저, cust 데이터프레임에 NaN null 결측치가 있는지 확인해 봅니다.
# 확인해 보니, 1번 인덱스에 Class 컬럼, sex컬럼, age 컬럼에 NaN null 결측치가 보이네요.

cust.head()
```

```python
# 위와 같이 cust 데이터프레임 데이터를 조회해서 NaN null 있음을 육안으로 확인할수도 있고
# info() 함수를 사용해서 확인할수도 있습니다.
# info() 함수의 결과를 보니, 총라인건수가 9,930건인데 
# class 컬럼, sex컬럼, age컬럼에 NaN null 존재하여 각각 8,529건, 9,249건, 9,178건으로 9,930건이 
# 안되는것을 확인할수 있습니다.

cust.info()
```

```python
# 결측치 확인할수 있는 다른 방법은 isnull() 함수를 통해서 NaN null 확인 할수 있습니다.
# 각 value 값들이 null 이냐?고 물어 보는겁니다. 
# null이 맞으면 True, null이 아니면 False로 표현합니다. 

cust.isnull().head()
```

```python
# isnull() 함수로 사용하니, 각 value 값들이 null 인지 , 아닌지 알수 있지만,
# 한눈에 안들어오므로, sum() 함수를 추가로 사용해서 컬럼별로 null 총합을 보도록 하겠습니다.
# class 컬럼에 1,401건의 null 값이 존재하고, sex 컬럼엔 681건, age 컬럼엔 752건의 null이 존재함을 알수 있습니다.

cust.isnull().sum()
```

```python
# fillna 함수를 사용해서 cust 데이터프레임 전체에 대해서 
# NaN null값을 숫자 15로 변경해 보겠습니다. 
# 여기서는 테스트이므로 임의의 값으로 변경하지만, 
# 실무에서는 일괄적으로 변경시 데이터의 편향이 틀어질수 있음에 유의해 주시기 바랍니다. 
# (버전 오류로 현재 3.14 버전)

fill_dict = {
    col: '15' if col in ['class', 'sex'] else 15
    for col in cust.columns
}
cust = cust.fillna(fill_dict)
```

```python
# 결측치를 숫자 15로 변경후 cust 데이터프레임을 확인해 보니
# 1번 인덱스에 존재하는 NaN도 15번 변경된것을 확인할수 있습니다.

cust.head()
```

```python
# 앞에서 info 함수를 통해 확인해 보니, 
# 모든 컬럼의 라인개수가 9930건으로  결측치가 1개도 없음을 알수 있습니다.

cust.info()
```

```python
# 다시 isnull() 함수를 사용해서 컬럼별 null 총합을 확인해 보겠습니다. 
# 모든 컬럼에는 null 갯수가 0으로 나와 null 없음을 확인할수 있습니다. 

cust.isnull().sum()
```

```python
# 결측치 채우기로 데이터가 변경되어 , 
# 다음 실습을 위해 다시 결측치가 있는 이전의 데이터프레임으로 복사하겠습니다. 

cust = cust_fix.copy()
```

```python
# cust 데이터프레임의 9927번 인덱스에 sex컬럼과 age 컬럼에 NaN null이 보이네요.
cust.tail()
```

fillna 함수의 Method 옵션을 사용해서, <br> 
데이터의 주변값으로 결측치를 채울수도 있습니다.

```python
# fillna 함수를 사용하여 결측치 채울때 특정값을 주었지만, 
# 이번에는 fillna 함수의 method 옵션을 'backfill' 주고 (구버전 1.x 이하, 현재 3.14 버전에는 이미 제거됨)
# 이번에는 bfill() 함수로,
# NaN 뒤에 있는 데이터값으로 결측치를 채워 보겠습니다.

# cust = cust.fillna(method='backfill') (구 버전)
cust = cust.bfill()
```

```python
# cust 데이터프레임의 뒤 3개로 보니, 
# 기존 9927 인덱스의 sex컬럼과 age 컬럼의 NaN null이 
# 바로 뒤의 데이터값 'F'와 40.0으로 변경되었습니다.

cust.tail()
```

```python
# 결측치 채우기로 데이터가 변경되어 , 
# 다음 실습을 위해 다시 결측치가 있는 이전의 데이터프레임으로 복사하겠습니다. 

cust = cust_fix.copy()
```

```python
# cust 데이터프레임의 9927번 인덱스에 sex컬럼과 age 컬럼에 NaN null이 보이네요.
cust.tail()
```

```python
# 이번에는 fillna 함수의 method 옵션을 'ffill' 주고  (구버전 1.x 이하, 현재 3.14 버전에는 이미 제거됨)
# NaN 앞에 있는 데이터값으로 결측치를 채워 보겠습니다. 

# cust = cust.fillna(method='ffill') (구 버전)
cust = cust.ffill()
```

```python
# 적용된 결과를 확인해 보니
# 9927번 인덱스의 sex컬럼과 age컬럼에 Nan Null이 바로 앞의 데이터 'M'과 12.0 으로 
# 채워진것을 알수 있습니다.

cust.tail()
```

Replace 함수를 사용해도 결측치 채우기가 가능합니다.

```python
# 1. 다시 결측치 실습을 위해 copy() 함수로 복사하고 
# 2. age 컬럼의 중앙값 확인 : 숫자 49.0으로 나오네요.

cust = cust_fix.copy()
cust[ 'age' ].median()
```

```python
# 3. replace()함수로 NaN null결측치를 age컬럼의 중앙값으로 채워 보겠습니다.
cust['age'] = cust['age'].replace( np.nan, cust['age'].median() )
```

```python
# 적용 결과를 확인해 보니, 1번 인덱스의 age 컬럼값이 숫자 49.0으로 변경되었습니다.
cust.head()
```

interpolate() 함수로 결측치 채우기     
선형방식은 값들을 같은 간격으로 처리하게된다.

```python
# 1. 다시 결측치 실습을 위해 copy() 함수로 복사하고 
# 2. 결측치가 존재하는 앞과 뒤의 값을 보고 그 중간의 값으로 대체하는 
#    interpolate 함수을 사용하여 결측값을 채워 보겠습니다.

cust = cust_fix.copy()
numeric_cols = cust.select_dtypes(include=['int64', 'float64']).columns # 타입 에러를 위해 수정
cust[numeric_cols] = cust[numeric_cols].interpolate()
```

```python
# interpolate 함수 적용 결과를 보니 
# 1번 인덱스의 age 컬럼의 NaN null값의 앞과 뒤의 값의 중간값
# 즉, 앞에는 28.0 이고 뒤에는 24.0 이므로, 여기의 중간값으로 26.0으로 
# 결측치를 채운것을 확인할수 있습니다. 

cust.head()
```

### <b>결측치 제거하기</b>
결측치를 채우기로 처리하다보면 경우에 따라 값의 왜곡이 발생하는 경우가 있습니다.    
이때는 제거하기 기법을 사용하여 결측치를 처리하면 데이터의 정합성을 보존 할 수 있습니다.

* listwise 방식 : record의 항목 중 1개의 값이라도 NaN이면 해당 데이터 행 전체 제거
* pairwise 방식 : 모든 항목이 NaN인 데이터 행만 제거

```python
# 결측치 실습을 위해 copy() 함수로 복사하겠습니다.
cust = cust_fix.copy()
```

```python
# 결측치 제거하기 실습을 하기전에 , 다시 info 함수로 cust 데이터프레임 구성 보겠습니다.
# 전체 라인은 9930 개이고, class, sex, age 컬럼에 결측지 존재함을 알수 있습니다.

cust.info()
```

```python
# listwise 방식 : 1개의 NaN이 존재하면 제거 하기
# dropna() 함수를 이용해서 1개의 NaN null이 있는 라인을 삭제해 보겠습니다.

cust = cust.dropna()
```

```python
# 앞쪽의 dropna 함수를 적용하고 info 함수로 cust 데이터프레임을 확인해 보니
# 1개라도 NaN null 값이 있는 라인을 삭제하여 
# 전체 라인이 9930개에서 8228개로 줄어듬을 확인할수 있습니다.

cust.info()
```

```python
# pairwise 방식 : 전체 라인이 모든 컬럼의 값이 NaN null일 경우에 해당 라인 제거하기
# 1. 다시 결측치 실습을 위해 copy() 함수로 복사하고 
# 2. dropna() 함수의 how 옵션을 'all'로 설정하고 수행하면 
#    전체 라인이 NaN null로 되어 있는 라인를 제거하게 됩니다. 

cust = cust_fix.copy()
cust = cust.dropna(how='all')
```

```python
# pairwise 방식으로 dropna 함수를 실행후 cust 데이터프레임 정보를 확인해 보니
# 기존 9930건와 똑같은 9930건으로 나옵니다. 
# 사실, 모든 컬럼의 값이 NaN null로 구성되어 있는 라인은 없습니다. 

cust.info()
```

특정 열에 있는 NaN만 참고하여 결측치를 제거하려면 Subset 파라미터를 사용하면 됩니다.

```python
# 특정열 안에서만 삭제하기
# 1. 다시 결측치 실습을 위해 copy() 함수로 복사하고 
# 2. cust 데이터프레임의 결측치 확인 보겠습니다. 

cust = cust_fix.copy()
cust.head()
```

```python
# 3. dropna() 함수의 subset 옵션을 사용해서 
#    'class' 특정컬럼에 대해서 NaN null 이 존재할 경우 해당 라인을 삭제해 보겠습니다.

cust = cust.dropna( subset=['class'] )
```

```python
# class 컬럼에 NaN null 있으면 해당 라인은 삭제하게 되어 
# cust 데이터프레임의 1번 인덱스가 삭제된것을 확인할수 있습니다.

cust.head()
```

### <b>결측치 처리한 데이터프레임을 파일로 저장하기</b>

```python
# 기존에 복사해둔 cust_fix 데이터프레임을 복사합니다. 
# 이제는 뒤쪽 실습에서 사용할 데이터프레임을 만들고 저장하도록 하겠습니다. 

cust = cust_fix.copy()
```

```python
# cust 데이터프레임 내용을 확인해 봅니다.
# 결측치 값들이 보입니다.

cust.head()
```

```python
# age 컬럼을 제외한 나머지 연속형 컬럼인 'service','avg_bill','A_bill','B_bill' 에 대해서만
# 결측치를 숫자 0으로 채우고 결과를 확인합니다.

cust[ ['service','avg_bill','A_bill','B_bill'] ] = cust[ ['service','avg_bill','A_bill','B_bill'] ].fillna(0)
cust.head()
```

```python
# 결측치가 있으면 해당 라인을 삭제합니다. 
# 위의 cust 데이터프레임과 dropna 결과후 cust 데이터프레임을 비교해 보니
# 1번 인덱스가 삭제된것을 알수 있습니다.

cust = cust.dropna()
cust.head()
```

```python
# 인덱스 초기화하여 새롭게 인덱스 만들겠습니다. 
# 이빨빠진 기존 인덱스는 버리도록 drop=True 옵션을 주고, 
# inplace=True 옵션을 사용해서 해당 결과를 다시 cust 데이터프레임에 적용하겠습니다.

cust.reset_index(drop=True, inplace=True)
cust
```

```python
# 다음 실습을 위해, 현재의 cust 데이터프레임을 'cust_data.csv' 파일로 저장하겠습니다.

cust.to_csv('cust_data.csv', index=False)
```
