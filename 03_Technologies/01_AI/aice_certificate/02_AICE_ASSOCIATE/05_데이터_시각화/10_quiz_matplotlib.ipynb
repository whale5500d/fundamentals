```python
import matplotlib.pyplot as plt
```

```python
# 주간 최고기온 그래프 그리기
x = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y = [28,30,29,31,32,31,31]

plt.figure(figsize=(15,4))
plt.plot(x, y)
plt.show()
```

```python
# 주간 최고기온 그래프 그리기 2
x = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y = [28,30,29,31,32,31,31]
x레이블 = 'Day'
y레이블 = 'Temp'

plt.figure(figsize=(15,4))
plt.plot(x, y)
plt.xlabel(x레이블)
plt.ylabel(y레이블)
plt.show()
```

```python
# 주간 최고기온 그래프 그리기 3
x = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y = [28,30,29,31,32,31,31]
x레이블 = 'Day'
y레이블 = 'Temp'
타이틀 = 'High Temperature'

plt.figure(figsize=(15,4))
plt.plot(x,y)
plt.xlabel(x레이블)
plt.ylabel(y레이블)
plt.title(타이틀)
plt.show()
```

```python
# 2개의 차트 그래프 그리기
x1 = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y1 = [28,30,29,31,32,31,31]
x2 = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y2 = [31,31,32,31,29,30,28]
x레이블 = 'Day'
y레이블 = 'Temp'
타이틀 = 'High Temperature'

plt.figure(figsize=(15,4))
plt.plot(x1, y1, label='this week')
plt.plot(x2, y2, label='last week')
plt.xlabel(x레이블)
plt.ylabel(y레이블)
plt.title(타이틀)
plt.legend()
plt.show()
```

```python
# 마커 사용하기
    # https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html
    # color: b, g, r, y
    # marker
        # ^: 삼각형
        # v: 역삼각형
        # o: 동그라미
        # s: 사각형
        # .: 점 (동그라미보다 작음)
    # line style
        # '-': 실선
        # '--': 구분선
        # ':': 점선
        # '-.': 점 + 구분선
x = list(range(0,10))
y1 = list(range(0,10))
y2 = list(range(0,20,2))
y3 = list(range(0,40,4))

plt.figure(figsize=(15,4))
plt.plot(x,y1, 'r^-')
plt.plot(x,y2, 'bs:')
plt.plot(x,y3, 'yo-.')
```

```python
# 여러 개의 차트 한 번에 그리기
x1 = [1,2,3]
y1 = [110,130,120]
x2 = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
y2 = [28,30,29,31,32,31,31]
x3 = range(len(y3))
y3 = [5, 3, 7, 10, 9, 5, 3.5, 8]

plt.figure(figsize=(15,5))

plt.subplot(131)
plt.plot(x1, y1)

plt.subplot(132)
plt.plot(x2, y2)
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('High Temperature')

plt.subplot(133)
plt.bar(x3, height=y3)

plt.show()
```

# 실습 2

- cust_data 데이터를 호출하여 질문에 답변하라.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../cust_data.csv')[:30]
```

```python
# 1. 평균 요금, A요금, B요금이 어떻게 변화되는지 선 그래프로 그려라.
    # 평균 요금: 빨간색, 네모 마커, 실선 스타일 사용
    # A 요금: 파랑색, 삼각형 마커, 점선 스타일 사용
    # B 요금: 노랑색, 역삼각형 마커, 점 + 구분선 스타일 사용
plt.figure()
plt.plot(df['avg_bill'], 'rs-')
plt.plot(df['A_bill'], 'b^:')
plt.plot(df['B_bill'], 'yv-.')
plt.show()
```

```python
# 2. 고객의 나이 분포를 표현하라.
    # 20개의 구간으로 구분
    # x축은 age, y축은 customer count 이름으로 설정
plt.figure()
plt.hist(df['age'], bins=20)
plt.xlabel('age')
plt.ylabel('customer count')
plt.show()
```

```python
# 3. 나이와 전체 요금 간의 관계를 산점도 그래프로 표현하라.
    # 그래프의 크기 (15, 5)로 설정
plt.figure(figsize=(15,5))
plt.scatter(df['age'], df['avg_bill'])
plt.show()
```
