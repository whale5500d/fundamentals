- cust_data.csv 데이터를 활용해 질문에 대답하라.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../cust_data.csv')[:300]
df.head()
```

```python
# 1. 아래 가이드를 따라서, 2개의 seaborn 산점도 그래프를 그려라.
    # 평균 요금과 나이 간의 산점도 그리기
    # 완성된 산점도를 색상 옵션을 사용해 등급별로 색상을 구분하라.

plt.figure()
sns.scatterplot(data=df, x='age', y='avg_bill')
plt.show()

plt.figure()
sns.scatterplot(data=df, x='age', y='avg_bill', hue='class')
plt.show()
```

```python
# 2. seaborn 히스토그램을 사용해 나이 분포를 확인하라.
    # 20개 구간으로 나누고
    # x축에는 'age', y축에는 'customer count'로 축 이름 설정하기
plt.figure()
sns.histplot(data=df, x='age', bins=20)
plt.xlabel('age')
plt.ylabel('customer count')
plt.show()
```

```python
# 3. 연속형 컬럼에 대해 상관관계를 구하라.
    # 상관관계에는 숫자가 보여야 한다.
    # 색상은 red로 설정하라.
df.info() # 연속형 컬럼 - age, service, avg_bill, A_bill, B_bill
continuous_columns = ['age', 'service', 'avg_bill', 'A_bill', 'B_bill']
continuous_df = df[continuous_columns].corr()

plt.figure()
sns.heatmap(data=continuous_df, annot=True, cmap='Reds')
plt.show()
```

```python
# 4. 위 3개의 그래프를 1행 3열에 보이도록 출력하라.
plt.figure(figsize=(15,5))

plt.subplot(131)
sns.scatterplot(data=df, x='age', y='avg_bill', hue='class')

plt.subplot(132)
sns.histplot(data=df, x='age', bins=20)
plt.xlabel('age')
plt.ylabel('customer count')

plt.subplot(133)
sns.heatmap(data=continuous_df, annot=True, cmap='Reds')

plt.show()
```
