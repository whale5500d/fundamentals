```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
df = pd.read_csv('../cust_data.csv')[:300]
df.head()
```

```python
# seaborn 목적별 차트 그리기

## 1. 산점도
# sns.scatterplot(data=df, x='age', y='avg_bill')
sns.scatterplot(data=df, x='age', y='avg_bill', hue='class')
plt.show()

## 2. 막대 그래프
# sns.barplot(data=df, x='class', y='age')
sns.barplot(data=df, x='class', y='age', hue='sex')
plt.show()

## 3. 추세선(회귀선)
    # 산점도에 회귀선을 그려서 어느 추세인지 확인할 수 있다.
# sns.lmplot(data=df, x='avg_bill', y='B_bill')
sns.lmplot(data=df, x='avg_bill', y='B_bill', hue='sex')
plt.show()

## 4. 항목별 개수 그래프
    # (그래프 색깔 바꾸기) https://seaborn.pydata.org/tutorial/color_palettes.html
# sns.countplot(data=df, x='class')
# sns.countplot(data=df, x='class', hue='sex')
# sns.countplot(data=df, y='class', hue='sex') # 가로 막대 그래프
sns.countplot(data=df, y='class', hue='sex', palette='spring')
plt.show()

## 5. 히스토그램 그래프
# sns.histplot(data=df, x='age')
# sns.histplot(data=df, x='age', bins=20)
sns.histplot(data=df, x='age', bins=20, hue='sex')
plt.show()

## 5. 산점도와 히스토그램을 한 번에 보여주는 상관관계 그래프\
# sns.jointplot(data=df, x="avg_bill", y="age")
sns.jointplot(data=df, x="avg_bill", y="age", kind='hex')
plt.show()

## 6. 연속형 데이터의 상관관계 그래프
    # ⚠️ 모든 데이터가 continous해야 한다. (연속형 데이터만 가능)
    # corr로 상관관계를 먼저 측정한다.
df_num = df.select_dtypes('number')
df_num.corr() # avg_bill, B_bill 칼럼 간 0.6으로 높은 상관관계 확인
# sns.heatmap(df_num.corr())
# sns.heatmap(df_num.corr(), annot=True)
sns.heatmap(df_num.corr(), annot=True, cmap='RdBu_r') # _r는 reverse로 1에 가까울수록 Red, -1에 가까울수록 Blue로 표시
plt.show()

## 7. 박스 차트
# sns.boxplot(data=df, y='avg_bill')
sns.boxplot(data=df, x='class', y='avg_bill')
plt.show()

## 8. 밀집도 차트
# sns.violinplot(data=df, y='avg_bill')
sns.violinplot(data=df, x='class', y='avg_bill')
plt.show()
```
