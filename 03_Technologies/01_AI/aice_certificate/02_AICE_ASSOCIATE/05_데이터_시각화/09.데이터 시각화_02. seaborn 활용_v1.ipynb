# **데이터 시각화 - seaborn 활용**

- 데이터 시각화는 데이터 분석과정에서 매우 중요한 부분입니다.     
- 프로젝트 초기 단계 또는 리포팅 단계에서 탐색적 데이터 분석(EDA) 결과를 공유할 때 유용합니다.    
- 데이터를 시각화 하면 값의 분포나 값 사이의 관계를 확인할수 있으며
- 수치 데이터에서 한눈에 파악하기 힘든 흐름, 추세 그리고 인사이트를 발견할수 있습니다.

## **1. Seaborn 라이브러리**

### <b>Seaborn 개요

- Seaborn은 Matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지입니다.    
- 기본적인 시각화 기능은 Matplotlib 패키지에 의존하게 됩니다. 
- Seaborn에 대한 자세한 내용은 웹사이트(http://seaborn.pydata.org/ )를 참조하세요.

### <b>Seaborn 라이브러리 불러오기

```python
# warnings 라이브러리 사용하여 warning 에러 출력하지 않도록 하겠습니다.

import warnings
warnings.filterwarnings('ignore')
```

```python
# seaborn 라이브러리 불러오기
import seaborn as sns
```

```python
# matplotlib 라이브러리 같이 불러옵니다. 
import matplotlib.pyplot as plt
```

### <b>사용할 데이터 불러오기

```python
# 우리가 사용할 데이터 불러오겠습니다. 

import pandas as pd

df = pd.read_csv("../cust_data.csv")[:300]
df.head()
```

<br>

## **2. Seaborn 목적별 차트 그리기**

### <b> scatterplot     
앞서 배웠듯 산점도는 두 값간의 상관관계를 확인해 볼 수 있습니다.

```python
# avg_bill(평균요금)과 age(나이)간의 상관관계를 확인해 봅시다.

sns.scatterplot(data=df, x='age', y='avg_bill' )
```

```python
# avg_bill(평균요금)과 age(나이)간의 상관관계를 확인해 봅시다.
# 추가로 , 색상(hue) 옵션을 사용하여 class(등급) 별로 색상 구분해 보겠습니다. 

sns.scatterplot(data=df, x='age', y='avg_bill', hue='class')
```

### <b> barplot
barplot은 x축엔 범주형 컬럼을 사용하고 y축엔 연속형 컬럼을 사용하여 범주별로 연속형컬럼의 바챠트 형식으로 보여줍니다. 

```python
# class(등급)별로 age(나이)에 대한 바챠트 그리기

sns.barplot(data=df, x='class', y='age', hue='class' )
plt.show()
```

```python
# class(등급)별로 age(나이)에 대한 바챠트 그리기
# 추가로 , 색상(hue) 옵션으로 성별('sex')로 구분해 보겠습니다. 

sns.barplot(data=df, x='class', y='age', hue='sex' )
plt.show()
```

### <b> lmplot
lmplot은 산점도에 회귀선을 그려주어 어느 추세인지 확인할수 있습니다.

```python
#lmplot을 사용하여 avg_bill과 B_bill의 상관관계를 확인하고 추세 그려 보겠습니다. 

sns.lmplot(data=df, x='avg_bill', y='B_bill' )
plt.show()
```

```python
# 색상(hue) 옵션으로 성별에 따라 어떻게 달라지는지 구분하기 
# 남자와 여자 추세를 비교해 보니, 여자가 남자보다 더 많이 요금을 내는 것으로 보입니다. 

sns.lmplot(data=df, x='avg_bill', y='B_bill',  hue='sex')
plt.show()
```

### <b> countplot
항목 별 갯수를 확인 할 때 사용합니다.

```python
# class(등급)별로 분포를 확인하기 
# D 클래스의 분포가 제일 많습니다. 

sns.countplot( data=df, x="class" )
plt.show()
```

```python
# class(등급) 분포를 색상(hue) 옵션을 사용해서 성별 ('F', 'M')로 나누어 보기

sns.countplot(data=df, x="class", hue="sex" )
plt.show()
```

```python
# 연속형 변수인 age컬럼에 대해 히스토그램 그래프 그리기 

sns.histplot(data=df, x='age', bins=10)
plt.show()
```

```python
# 연속형 변수인 age컬럼에 대해 히스토그램 그래프 그리기  
# 추가로 20개로 구간화 나누고, 색상(hue) 옵션을 활용하여 성별(sex)로 구분해 보겠습니다. 

sns.histplot(data=df, x='age', bins=20, hue='sex')
plt.show()
```

```python
#성별 고객등급 분포를 가로 막대 그래프로 확인하기
sns.countplot(data=df, y="class", hue="sex",)
plt.show()
```

#### 그래프 색깔 바꾸기   
palette값을 설정하면 원하는 톤의 그래프를 그릴 수 있습니다.    
더 많은 palette색상은 담은을 참고하세요.      https://seaborn.pydata.org/tutorial/color_palettes.html

```python
# palette값을 spring로 설정하여 색을 변경하기

sns.countplot(data=df, y="class", hue="sex",  palette='spring')
plt.show()
```

### <b> jointplot
산점도와 히스토그램을 한번에 보여주어 데이터의 분포와 상관관계를 한번에 볼 수 있습니다.

```python
# jointplot을 사용하여 avg_bill(평균요금)과 age(나이)간의 관계 확인하기

sns.jointplot(x="avg_bill", y="age", data=df)
plt.show()
```

kind 옵션을 사용하면 산점도 영역의 그래프를 다른 그래프로 변경 할 수 있습니다

```python
#kind 옵션을 사용하여 다양한 그래프를 그리기

sns.jointplot(x="avg_bill", y="age", data=df, kind='hex')
plt.show()
```

### <b> Heatmap (상관관계)
상관관계를 확인하기 위해서 주의해야 할 점은 모든 데이터가 'continous'해야 한다는 점입니다.     
corr()은 상관관계를 측정 할 수 있는 함수입니다.

```python
# 연속형 변수만을 뽑아 df_num 데이터프레임 만들기

df_num = df.select_dtypes('number')
df_num.head()
```

```python
# df_num 데이터프레임에 대해서 상관관계 확인하기
# avg_bill(평균요금) 컬럼과 B_bill 컬럼간의 0.6으로 높은 상관관계를 보이고 있습니다. 

df_num.corr()
```

 heatmap을 사용하면 두개의 카테고리 값에 대한 값변화를 한 눈에 알기 쉽기 때문에   
 두 값의 상관관계를 나타낼때 주로 사용 합니다.

```python
# 컬럼별 상과관계를 heatmap 그래프로 그리기
# annot=True 옵션을 사용하여 상관관계 숫자 표현하기
# 히트맵 색상을 cmap 옵션으로 활용해서 더 보기 좋게 보이기 

sns.heatmap(df_num.corr(), annot=True, cmap='RdBu_r')
plt.show()
```

### <b>boxplot
수치적 자료를 표현하는 그래프입니다.      
수치적 자료로부터 얻어낸 통계량인 5가지 요약 수치(Lower Fence, 제 1사분위값, 제 2사분위값, 제 3사분위값, Upper Fence)를 가지고 그리며,     
전처리에서 배웠던 describe()함수와 함께 데이터 분포를 확인하는데도 유용하게 사용할수있습니다.     
또한, Lower Fence와 Upper Fence를 범위를 벗어나는 이상치도 확인할수 있습니다. 

```python
# class(등급)별로 avg_bill(평균요금)에 대한 boxplot 그리기 
# upper fence와 lower fence 벗어난 이상치가 보이고 있습니다. 

sns.boxplot( data=df, x="class", y='avg_bill' )
plt.show()
```

### <b>violinplot
violinplot은 Boxplot과 모양은 비슷하지만 밀집도를 함께 볼 수 있어 데이터 탐색에 유용하게 사용할 수 있습니다.

```python
# 등급('class')별로 avg_bill(평균요금)에 대한 violinplot을 그리기

sns.violinplot( data=df, x='class', y='avg_bill' )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../cust_data.csv")[:300]
df.head()
```

```python
# 1. 아래 가이드에 따라서, 2개의 seaborn 산점도 그래프를 그리세요.
#    * avg_bill(평균요금)과 age(나이)간의 산점도를 그리세요

sns.scatterplot( data=df, x='age', y='avg_bill'  )
plt.show()
```

```python
# 1. 아래 가이드에 따라서, 2개의 seaborn 산점도 그래프를 그리세요.
#    * avg_bill(평균요금)과 age(나이)간의 산점도를 그리고, 색상(hue) 옵션을 사용해서 등급('class') 별로 색상을 구분해 보세요.

sns.scatterplot( data=df, x='age', y='avg_bill' , hue='class' )
plt.show()
```

```python
# 2. searborn 히스트그램을 사용하여 고객의 나이 분포를 확인해 보세요.
#    * 20개 구간으로 나누고, X축에는 'age', Y축에는 'customer count'로 축이름을 설정합니다.

sns.histplot(data=df, x='age', bins=20)
plt.xlabel('age')
plt.ylabel('customer count')
plt.show()
```

```python
# 3. [ 'age', 'avg_bill', 'A_bill', 'B_bill' ] 연속형 컬럼들에 대해서 상관관계를 구하고 히트맵을 그리세요.
#     * 상관관계 숫자가 보이도록 annot=True 하세요.
#     * 히트맵 색상을 cmap='Reds'로 설정하세요.

#  [ 'age', 'avg_bill', 'A_bill', 'B_bill' ]  연속형 변수에 대한 상관관계 구해 보기
df[ [ 'age', 'avg_bill', 'A_bill', 'B_bill' ]  ].corr()
```

```python
# 히트맵 그리기
#     * 상관관계 숫자가 보이도록 annot=True 하세요.
#     * 히트맵 색상을 cmap='Reds'로 설정하세요.

sns.heatmap( df[ [ 'age', 'avg_bill', 'A_bill', 'B_bill' ]  ].corr(), annot=True, cmap='Reds' )
plt.show()
```

```python
# 산점도 : sns.scatterplot( data=df, x='age', y='avg_bill' , hue='class' )
# 히스토그램 : sns.histplot(data=df, x='age', bins=20)
# 히트맵 : sns.heatmap( df[ [ 'age', 'avg_bill', 'A_bill', 'B_bill' ]  ].corr(), annot=True, cmap='Reds' )

plt.figure( figsize=(15,5) )

plt.subplot(131)
sns.scatterplot( data=df, x='age', y='avg_bill' , hue='class' )

plt.subplot(132)
sns.histplot(data=df, x='age', bins=20)

plt.subplot(133)
sns.heatmap( df[ [ 'age', 'avg_bill', 'A_bill', 'B_bill' ]  ].corr(), annot=True, cmap='Reds' )

plt.show()
```
