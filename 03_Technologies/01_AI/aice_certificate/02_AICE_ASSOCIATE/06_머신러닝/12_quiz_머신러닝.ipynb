- 아래 가이드에 따라 tips 데이터프레임을 활용하여 머신러닝 모델을 만들어라.

```python
import seaborn as sns

row_tips = sns.load_dataset('tips')

row_tips.info() # sex, smoker, day, time은 category dtype이므로, 숫자형 데이터로 변형 필요

# 데이터 전처리
## 1. 직접 처리
row_tips['sex'] = row_tips['sex'].map({'Female':0, 'Male':1})
row_tips['smoker'] = row_tips['smoker'].map({'No':0, 'Yes':1})
row_tips['time'] = row_tips['time'].map({'Lunch':0, 'Dinner':1})

## 2. Feature Engineering
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
row_tips['day'] = le.fit_transform(row_tips['day'])

 # 결과 확인 및 저장
row_tips
row_tips.to_csv('tips2.csv', index=False)
```

```python
# 회귀 모델
    # tips 데이터프레임 사용

## 1. x, y 데이터 만들기
    # y: total_bill, x: 그 외
import pandas as pd

tips = pd.read_csv('tips2.csv')

y = tips['total_bill']
x = tips.drop(columns='total_bill')

## 2. 학습 데이터, 평가 데이터 만들기
    # 학습:평가 = 8:2
    # 렌덤값 41
from sklearn.model_selection import train_test_split # ⭐️ model_selection 까먹음
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)

# values 값을 바로 사용하기 위해 데이터프레임을 넘파이로 변경
x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values

## 3. KNN 모델 사용
    # 이웃 개수를 3으로 조정
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(x_train, y_train)
knr.score(x_test, y_test) # 0.31038693219889857

## 4. 예측하기
    # 시뮬레이션 데이터: [3.23, 1. , 0. , 2. , 1. , 2. ]
# x_test[0], y_test[0] # (array([3.06, 1.  , 0.  , 2.  , 1.  , 2.  ]), 13.94)
simul = [3.23, 1. , 0. , 2. , 1. , 2. ]
pred = knr.predict([simul]) # ⭐️ 이차원 배열 사용해야
print(pred) # [20.00333333]
```

```python
# 분류 모델
    # tips 데이터프레임 사용

## 1. x, y 데이터 만들기
    # y: day, x: 그 외
import pandas as pd

tips = pd.read_csv('tips2.csv')

y = tips['day']
x = tips.drop(columns='day')

## 2. 학습 데이터, 평가 데이터 만들기
    # 학습:평가 = 8:2
    # 렌덤값 41
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)

x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values

## 3. KNN 모델 사용
    # 이웃 개수 : 3
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=3)
knc.fit(x_train, y_train)
knc.score(x_test, y_test) # 0.5306122448979592

## 4. 예측하기
    # 시뮬레이션 데이터: [14.78,  3.23,  1.  ,  0.  ,  1.  ,  2.  ]
simul = [14.78,  3.23,  1.  ,  0.  ,  1.  ,  2.  ]
pred = knc.predict([simul])
print(pred) # [1]
```
