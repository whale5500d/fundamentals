# 실습 5

- 3반 5명의 학생의 국어, 영어 성적에 대한 딕셔너리 데이터를 통해 국어, 영어의 각 합과 평균을 구하라.
```
dic = {
   '국어' : [70, 95, 80, 80, 100] ,
   '영어' : [65, 95, 80, 70, 95]  
}
```

````python
dic = {
   '국어' : [70, 95, 80, 80, 100] ,
   '영어' : [65, 95, 80, 70, 95]  
}

ko_sum = 0
ko_length = len(dic['국어'])
for ko_score in dic['국어']:
    ko_sum += ko_score

print('국어 합: ', ko_sum)
print('국어 평균: ', ko_sum / ko_length)

en_sum = 0
en_length = len(dic['영어'])
for en_score in dic['영어']:
    en_sum += en_score

print('영어 합: ', en_sum)
print('영어 평균: ', en_sum / en_length)
````

# 실습 8

- 아래 20개의 난수로 이뤄진 리스트에 대해 질문에 답변하라.
- [37, 570, 606, 646, 354, 603, 216, 102, 362, 588, 490, 494, 657, 820, 571, 33, 53, 936, 313, 38]

````python
# 1. 리스트를 list 변수에 담아 출력하라.
list = [37, 570, 606, 646, 354, 603, 216, 102, 362, 588, 490, 494, 657, 820, 571, 33, 53, 936, 313, 38]
list
````

````python
# 2. 위 리스트에서 슬라이싱을 사용해서 5번째 위치부터 끝까지 리스트를 만들어라.
list2 = list[4:]
list2
````

````python
# 3. 만든 리스트의 크기를 출력하라.
len(list2)
````

````python
# 4. 16개의 리스트를 정렬하고 최소값과 최대값을 구하라.
list2.sort()
list2
min(list2)
max(list2)
````

````python
# 5. 정렬된 리스트에 대해 홀수와 짝수로 구분하여 리스트로 저장하고 각 합과 평균을 구하라.
evenList = []
oddList = []

for i in list2:
    if i % 2 == 0:
        evenList.append(i)
    else:
        oddList.append(i)

evenSum = sum(evenList)
evenLength = len(evenList)

print('even sum: ', evenSum)
# print('even length: ', evenLength)
print('even average: ', evenSum / evenLength)

oddSum = sum(oddList)
oddLength = len(oddList)

print('odd sum: ', oddSum)
# print('odd length: ', oddLength)
print('odd average: ', oddSum / oddLength)
````

# 실습 9

- 학생 10명의 국어, 영어, 수학 점수표로 질문에 답변하라.
```
DIC = {
    '국어' : [80, 85, 90, 100, 90, 95, 80, 70, 85, 75],
    '영어' : [90, 90, 95, 85, 80, 85, 70, 75, 90, 85],
    '수학' : [60, 80, 70, 75, 90, 80, 75, 65, 85, 80]
}
```

````python
# 1. DIC 딕셔너리에 정의하라.
DIC = {
    '국어' : [80, 85, 90, 100, 90, 95, 80, 70, 85, 75],
    '영어' : [90, 90, 95, 85, 80, 85, 70, 75, 90, 85],
    '수학' : [60, 80, 70, 75, 90, 80, 75, 65, 85, 80]
}
DIC
````

````python
# 2. 국어, 영어, 수학의 총합과 평균을 구하라.

for subject in DIC.keys():
    list = DIC[subject]
    print('sum: ', sum(list))
    print('average: ', sum(list) / len(list))
````

````python
# 3. 's총합'과 's평균' 리스트에 국어, 영어, 수학의 점수를 추가하라.

s총합 = []
s평균 = []

for subject in DIC.keys():
    subjectList = DIC[subject]
    s총합.append(sum(subjectList))
    s평균.append(sum(subjectList) / len(subjectList))
````

````python
# 4. 's총합'과 's평균' 리스트를 for문으로 출력하라.

for i in s총합:
    print(i)
for i in s평균:
    print(i)
````
