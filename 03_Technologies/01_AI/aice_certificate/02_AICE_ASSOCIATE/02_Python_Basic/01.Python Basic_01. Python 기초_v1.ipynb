## __5. 파이썬 조건문__
- 조건절을 사용하여 조건절이 참이면 아래 수행문을 수행하고 거짓이면 아래 수행문을 수행하지 않는다.
- 문법 : 
         if 조건절 : 
            수행문
         else :
            수행문
- 문법을 보시면 ':' 있는걸 확인할수 있고 들어쓰기도 반드시 되어야 제대로 수행된다.

```python
# 변수 money 를 정의 합니다.
money = 10000
```

```python
# money 변수 확인합니다.
money
```

```python
# 조건문 사용하기
# money 변수에 돈이 만원 이상이면 , "택시 타고 간다" 라고 출력하고
# 그렇지 않으면 "걸어간다" 라고 출력한다.

money = 10000

if  money >= 10000  :
    print('택시 타고 간다')
else :
    print('걸어간다')
```

```python
# money 변수를 5000으로 변경한 후 조건문 수행해 보자.

money = 5000

if  money >= 10000  :
    print('택시 타고 간다')
else :
    print('걸어간다')
```

## __6. 파이썬 반복문__
- for 문을 사용하여 수행문들을 여러번 수행합니다. 
- for 문법
         for 변수 in 리스트 : 
            수행문1
            수행문2

```python
for i in range(1, 6) :
    print('숫자를 {}을 입력합니다.'.format(i))

```

```python
# 아래 문자열을 숫자만 변경해 가면서 출력하려고 합니다.
# 숫자를 1을 입력합니다.
# 숫자를 2을 입력합니다.
# 숫자를 3을 입력합니다.
# 숫자를 4을 입력합니다.
# 숫자를 5을 입력합니다.
# 일일히 1줄씩 print() 함수로 출력하려니 , 너무 많은 수작업이 필요하네요.
# 쉽게 반복문을 처리해야 되겠네요. 
# for 문을 사용해서 간단하게 처리해 보시죠

for i in [ 1, 2, 3, 4, 5 ] :
    print('숫자를 {}을 입력합니다.'.format(i))
```

```python
# 앞에서 배운 range함수와 list 함수를 같이 이용 합니다.

for i in list( range(1, 6) ) :
    print('숫자를 {}을 입력합니다.'.format(i))
```

# **[실습] 파이썬 프로그램밍 코딩하기**
1. 1부터 200까지의 숫자로 된 리스트 만들기
2. 위의 리스트에서 조건문을 사용하여 홀수만 뽑아내서 모든 홀수의 합을 출력하세요.
3. 2단부터 9단까지의 구구단을 출력하세요.
4. 아래 3반 5명의 학생들의 국어, 영어 성적에 대한 딕셔너리 데이터를 가지고 국어와 영어의 합과 평균을 구하세요.
<br>
{
   '국어' : [70, 95, 80, 80, 100] ,
   '영어' : [65, 95, 80, 70, 95]  
}


```python
# 1. 1부터 200까지의 숫자로 된 리스트 만들기
# 200까지 하면 출력이 많이 나와 11까지로 변경해서 테스트 수행함

num_list = list(range(1,11))
num_list
```

```python
# 2. 위의 리스트에서 조건문을 사용하여 홀수만 뽑아내서 모든 홀수의 합을 출력하세요.

for i in num_list:
    print(i)
```

```python
sum = 0 
for i in num_list:
    if i % 2 == 1  :  # 홀수
        sum = sum + i
        print(i)
print(sum)
```

```python
# 3. 2단부터 9단까지의 구구단을 출력하세요.

for i in [2,3,4,5,6,7,8,9] :  # list(range(1,10))
    for j in [1,2,3,4,5,6,7,8,9] : 
        cal = i * j 
        print('{} X {} = {}'.format(i,j,cal))
```

```python
# 4. 아래 3반 5명의 학생들의 국어, 영어 성적에 대한 딕셔너리 데이터를 가지고 국어와 영어의 합과 평균을 구하세요.
# 먼저, grade 딕셔너리 정의 
grade = { '국어' : [70, 95, 80, 80, 100] , 
          '영어' : [65, 95, 80, 70, 95] }
```

```python
# grade 딕셔너리 확인
grade
```

```python
# grade 딕셔너리의 '국어' 키 접근
grade['국어']
```

```python
# grade 딕셔너리의 '영어' 키 접근
grade['영어']
```

```python
# 국어의 점수 합과 평균 구하기

sum = 0
for i in grade['국어'] : 
    # print(i)
    sum = sum + i

print('합 : {}'.format(sum))
print('평균 : {}'.format(sum / 5))
```

```python
# 양아의 점수 합과 평균 구하기

sum = 0
for i in grade['영어'] : 
    # print(i)
    sum = sum + i

print('합 : {}'.format(sum))
print('평균 : {}'.format(sum / 5))
```
