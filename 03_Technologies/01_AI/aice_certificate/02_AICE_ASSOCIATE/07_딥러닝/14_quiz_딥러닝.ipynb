```python
# 데이터 로드
import seaborn as sns

row_tips = sns.load_dataset('tips')
```

```python
# 데이터 전처리

## 1. 탐색

## 2. 처리(직접 처리)

## 3. 처리(Feature Engineering)

## 4. 저장
```

```python
# 데이터 전처리

## 1. 탐색
# row_tips.info() # sex, smoker, day, time는 dtype이 category
# row_tips.head()
# row_tips['time'].value_counts()

## 2. 처리(직접 처리)
row_tips['sex'] = row_tips['sex'].map({'Female':0, 'Male':1})
row_tips['smoker'] = row_tips['smoker'].map({'No':0, 'Yes':1})
row_tips['time'] = row_tips['time'].map({'Lunch':0, 'Dinner':1})

## 3. 처리(Feature Engineering)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
row_tips['day'] = le.fit_transform(row_tips['day'])

## 4. 저장
row_tips.to_csv('tips2.csv', index=False)
```

```python
# 분류 모델

## 1. x, y 데이터 만들기
    # y: total_bill, x: 그 외

## 2. 학습, 평가 데이터 만들기
    # 학습:평가 = 8:2
    # 랜덤값: 41

## 3. DNN 모델 생성
    # 첫째 히든: 유닛 6개
    # 둘째 히든: 유닛 4개
    # 히든 activation: relu
    # 아웃풋: 유닛 1개
    # 아웃풋 activation: linear

## 4. DNN 성능 향상 (최적화)
    # 최적화: loss='mse', optimizer='adam', metrics=['mse', 'mae']
    # EarlyStopping 중지 기점: 최대 3회
    # ModelCheckpoint

## 5. 학습 진행
    # epochs=50, batch_size=8

## 6. 모델 예측
```

```python
# 회귀 모델

## 1. x, y 데이터 만들기
    # y: total_bill, x: 그 외
import pandas as pd

tips = pd.read_csv('tips2.csv')
y = tips['total_bill']
x = tips.drop(columns='total_bill')

## 2. 학습, 평가 데이터 만들기
    # 학습:평가 = 8:2
    # 랜덤값: 41
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)

x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values

## 3. DNN 모델 생성
    # 첫째 히든: 유닛 6개
    # 둘째 히든: 유닛 4개
    # 히든 activation: relu
    # 아웃풋: 유닛 1개
    # 아웃풋 activation: linear
import tensorflow as tf # ⭐️
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
print(x_train.shape, y_train.shape)
model = Sequential()
model.add(Dense(6, input_shape=(6,)))
model.add(Dense(4))
model.add(Dense(1, activation=None))
model.summary()

## 4. DNN 성능 향상 (최적화)
    # 최적화: loss='mse', optimizer='adam', metrics=['mse', 'mae']
model.compile(loss='mse', optimizer='adam', metrics=['mse', 'mae'])

## 5. 학습 진행
    # EarlyStopping 중지 기점: 최대 3회
    # ModelCheckpoint
    # epochs=50, batch_size=8
    # history에 저장
es = EarlyStopping(patience=3, verbose=1)
mc = ModelCheckpoint('best_model.keras', save_best_only=True, verbose=1)
history = model.fit(x_train, y_train, epochs=50, batch_size=8, validation_data=(x_test, y_test), callbacks=[es, mc])

## 6. 모델 예측
    # 시뮬 데이터: [3.23, 1., 0., 2., 1., 2.]
import numpy as np
simul = [3.23, 1., 0., 2., 1., 2.]
test_input = np.array([simul], dtype='float32')
model.predict(test_input)
```

```python
# 분류 모델

## 1. x, y 데이터 만들기
    # y: smoker, x: 그 외
import pandas as pd
tips = pd.read_csv('tips2.csv')
y = tips['smoker']
x = tips.drop(columns='smoker')

## 2. 학습, 평가 데이터 만들기
    # 학습:평가 = 8:2
    # 랜덤값: 41
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)

x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values

## 3. DNN 모델 생성
    # 첫째 히든: 유닛 6개
    # 둘째 히든: 유닛 4개
    # 히든 activation: relu
    # 아웃풋: 유닛 1개
import tensorflow as pf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
# print(x_train.shape, y_train.shape)
model = Sequential()
model.add(Dense(6, activation='relu', input_shape=(6,)))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
## 4. DNN 성능 향상 (최적화)
    # 최적화: loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']
model.compile(loss='mse', optimizer='adam', metrics=['mse', 'mae'])
# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
## 5. 학습 진행
    # EarlyStopping 중지 기점: 최대 3회
    # ModelCheckpoint
    # epochs=50, batch_size=8
es = EarlyStopping(patience=3, verbose=1)
mc = ModelCheckpoint('best_model.keras', save_best_only=True, verbose=1)
history = model.fit(x_train, y_train, epochs=50, batch_size=8, validation_data=(x_test, y_test), callbacks=[es, mc])
```

```python

## 6. 모델 예측
    # 시뮬 데이터: [23.33, 5.65, 1., 2., 1., 2.]
import numpy as np
simul = [23.33, 5.65, 1., 2., 1., 2.]
test_input = np.array([simul], dtype='float32')
model.predict(test_input)
```
