data = {

````python
# 1. 해당 딕셔너리를 활용해 df 데이터프레임을 만들고 내용 조회하기
import pandas as pd
import numpy as np

data = {
          "지점": ["A", "A", "B", "B"],
          "상품": ["P1", "P2", "P1", "P2"],
          "Q1": [112, 134, 156, 178],
          "Q2": [212, 234, 256, 278],
          "Q3": [312, 334, 356, 378],
          "Q4": [412, 434, 456, 478]
}
df = pd.DataFrame(data)
df
````

````python
# 10. '지점' 컬럼 기준 그룹화하고 평균 구하기 ⭐️
# df.groupby(['지점'])[['Q1', 'Q2', 'Q3', 'Q4']].mean() # 숫자 타입만 선택하여 평균 구하기
df.drop('상품', axis=1).groupby('지점').mean() # 숫자 외 타입 제거하여 평균 구하기
````

````python
# 11. '상품' 컬럼 기준 그룹화하고 평균 구하기
df.drop('지점', axis=1).groupby('상품').mean()
````

````python
# 12. pivot_table 활용해 인덱스는 '지점', 컬럼은 '상품', values는 'Q1'으로 평균 구하기 ⭐️
# df.pivot_table(index='지점', columns='상품', values='Q1').mean() # 잘못 작성한 코드
# aggFunc='mean'이므로 mean()할 경우 한 번 더 평균 진행. mean()은 컬럼별 평균 구하는 함수
df.pivot_table(index='지점', columns='상품', values='Q1')
# pd.pivot_table(data=df, index='지점', columns='상품', values='Q1')
````

````python
# 13. df 데이터프레임 '지점' 컬럼을 인덱스로 만들고 df1에 할당하기
df1 = df.set_index('지점')
df1
````

````python
# 14. df1 데이터프레임 인덱스를 리셋하기
df1.reset_index()
````

````python
# 15. df 데이터프레임 '지점'과 '상품' 2개의 멀티인덱스를 만들고 df2에 저장
df2 = df.set_index(['지점', '상품'])
df2
````

````python
# 16. df2 데이터프레임에서 '상품' 인덱스를 컬럼 레벨로 올리기
df2.unstack('상품')
````

````python
# 17. df2 데이터프레임에서 '지점' 인덱스를 컬럼 레벨로 올리고 df3에 저장
df3 = df2.unstack('지점')
df3
````

````python
# 18. df3 데이터프레임에서 '지점' 컬럼을 인덱스 레벨로 내리기
df3.stack('지점')
````
