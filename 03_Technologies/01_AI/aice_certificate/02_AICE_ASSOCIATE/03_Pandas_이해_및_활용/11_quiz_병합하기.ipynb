# 실습 1

- 아래의 2개의 리스트를 활용하여 질문에 답하라.
```
data1 = [
         ['홍길동',23,'남'],
         ['이순신',55,'남']
]

data2 = [
         ['신사임당',48,'여'], 
         ['장보고',22,'남']
]
```

````python
import pandas as pd
import numpy as np

data1 = [
         ['홍길동',23,'남'],
         ['이순신',55,'남']
]

data2 = [
         ['신사임당',48,'여'], 
         ['장보고',22,'남']
]
# 컬럼명은 이름, 나이, 성별로 설정하기

df1 = pd.DataFrame(data1, columns=['이름', '나이', '성별'])
df2 = pd.DataFrame(data2, columns=['이름', '나이', '성별'])

# 1. df1, df2 데이터프레임을 위 아래로 합치기
pd.concat([df1, df2])
# 2. df1, df2 데이터프레임을 인덱스를 새로 만들어서 위 아래로 합치기
pd.concat([df1, df2], ignore_index=True)
````

# 실습 3

- 아래 리스트를 활용하여 질문에 답하라.
```
data1 = [
         ['홍길동',23,'남'],
         ['이순신',55,'남'], 
         ['신사임당',48,'여']
]

columns=['이름','나이','성별']

data2 = [
         ['홍길동','터미네이터',5],
         ['홍길동','비긴어게인',2], 
         ['홍길동','살인의추억',4], 
         ['신사임당','비긴어게인',5],
         ['장보고','살인의추억',3]
]

columns=['이름','나이','성별']
```

````python
import pandas as pd

data1 = [
         ['홍길동',23,'남'],
         ['이순신',55,'남'], 
         ['신사임당',48,'여']
]

columns1=['이름','나이','성별']

data2 = [
         ['홍길동','터미네이터',5],
         ['홍길동','비긴어게인',2], 
         ['홍길동','살인의추억',4], 
         ['신사임당','비긴어게인',5],
         ['장보고','살인의추억',3]
]

columns2=['이름','영화','평점']

df1 = pd.DataFrame(data=data1, columns=columns1)
df2 = pd.DataFrame(data=data2, columns=columns2)

# 1. df1 데이터프레임 보기
df1
# 2. df2 데이터프레임 보기
df2
# 3. 이름 컬럼을 key로 해서 교집합 merge (df3로 할당)
df3 = pd.merge(df1, df2, on='이름')
# 4. 이름 컬럼을 key로 해서 df1 기준 merge
pd.merge(df1, df2, on='이름', how='left')
# 5. 이름 컬럼을 key로 해서 df2 기준 merge
pd.merge(df1, df2, on='이름', how='right')
# 6. 이름 컬럼을 key로 해서 합집합 merge
# pd.merge(df1, df2, on='이름', how='outer')
df1.merge(df2, on='이름', how='outer')

# 7. df3에서 이름별, 영화별 평점의 합 구하기
# df3.groupby(['이름', '영화'])['평점'].sum() # 결과만 빠르게 알고 싶을 때
pd.pivot_table(data=df3, index='이름', columns='영화', values='평점', aggfunc=np.sum) # 결과를 표로 깔끔하게 보고싶을 때
````
