````python
import pandas as pd

# 1. 아래 딕셔너리를 활용하여 df 데이터프레임을 만들어라.
data = {
    '이름' : ['루피', '조로', '상디', '나미', '우솝', '쵸파', '로빈', '프랑키'],
    '고향' : ['Eastblue','Northblue','Northblue', 'Eastblue', 'Grandline', 'Grandline', 'OHARA', 'Waterseven'],
    '키' : [174, 181, 180, 170, 173, 90, 188, 240],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    '기술' : ['레드호크', '아수라', '디아블잠브', '', '', '', '풀루르', '라디컬 빔']   
}

df = pd.DataFrame(data)
df
````

````python
# 2. 앞쪽 5개 데이터 보기
df.head()

# 3. 뒤쪽 5개 데이터 보기
df.tail()

# 4. info 보기
df.info()

# 5. 통계정보 보기
df.describe()

# 6. shape 정보 보기
df.shape # (8, 9)

# 7. index 보기
df.index # RangeIndex(start=0, stop=8, step=1)

# 8. columns 보기
df.columns # Index(['이름', '고향', '키', '국어', '영어', '수학', '과학', '사회', '기술'], dtype='str')

# 9. columns의 마지막 2개의 컬럼만 뽑아내기 ⭐️
df.columns[-2:] # Index(['사회', '기술'], dtype='str')

# 10. values 보기
df.values

# 11. 키컬럼과 영어컬럼만 보기
df[['키', '영어']]
````

````python
# 12. 5번 인덱스의 '국어' 컬럼의 값 확인하기 ⭐️
df.loc[5, '국어'] # 80

# 13. 국어 점수가 80 이상인 데이터만 뽑아 조회
con_isAbove80inKorean = df['국어'] >= 80
df[con_isAbove80inKorean]

# 14. 국어 점수가 80 이상이고 수학 점수가 90 이상인 데이터만 뽑아 조회
con_isAbove90inMath = df['수학'] >= 90
df[con_isAbove80inKorean & con_isAbove90inMath]

# 15. 국어, 영어, 수학 3과목의 점수를 합쳐 '핵심과목' 컬럼에 저장
df['핵심과목'] = df['국어'] + df['수학'] + df['영어']
df

# 16. 국어, 영어, 수학 3과목의 점수로 '핵심평균' 컬럼에 저장
df['핵심평균'] = df['핵심과목'] / 3
df

# 17. 이름 컬럼 삭제하기 ⭐️
df.drop('이름', axis=1)
````
